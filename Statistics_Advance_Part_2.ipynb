1. **What is hypothesis testing in statistics?**

  Ans. In statistics, hypothesis testing is a formal procedure used to determine whether there is enough evidence in a sample of data to infer that a certain condition is true for the entire population. It's a fundamental concept that allows us to make decisions or draw conclusions about a population based on sample data.
---

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**

  Ans. The null hypothesis (H0) is a statement of no effect, no difference, or no relationship. It represents the status quo or a position of skepticism. It's what we initially assume to be true about the population. The goal of hypothesis testing is often to try and disprove this statement.

  The alternative hypothesis (Ha or H1), on the other hand, is a statement that contradicts the null hypothesis. It proposes that there is an effect, a difference, or a relationship. It's what the researcher often wants to find evidence for.
---

3. **What is the significance level in hypothesis testing, and why is it important?**

  Ans. The significance level (α) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true. In other words, it's the probability of making a Type I error (a false positive).

  - It's a pre-determined value chosen by the researcher before conducting the hypothesis test.
  - Commonly used significance levels are 0.05 (5%) and 0.01 (1%).
  - If we set α=0.05, it means there's a 5% chance that we will reject the null hypothesis even if it's correct.

  The significance level is important for the following reasons:

  - Sets a Threshold for Decision Making: The significance level acts as a cut-off point for deciding whether the evidence from the sample data is strong enough to reject the null hypothesis. We compare the p-value (the probability of observing our data or more extreme data if the null hypothesis were true) to α.
      - If the p-value ≤α, we reject the null hypothesis.
      - If the p-value >α, we fail to reject the null hypothesis.
  - Controls the Probability of a Type I Error: By setting α, we explicitly control the risk of incorrectly concluding that there is an effect or a difference when, in reality, there isn't one in the population. This is crucial in fields where false positives can have significant consequences (e.g., medical research, where incorrectly claiming a drug is effective could harm patients).
  - Provides a Standard for Statistical Significance: The significance level provides a common benchmark for determining whether a result is considered "statistically significant." If a study reports results as significant at the 0.05 level, it implies that the p-value was less than or equal to 0.05, meaning the observed data were unlikely to have occurred by random chance alone if the null hypothesis were true.
---

4. **What does a P-value represent in hypothesis testing?**

  Ans. The P-value in hypothesis testing represents the probability of obtaining test results at least as extreme as the results actually observed, assuming the null hypothesis is true.

  For Example:

  - A small P-value (typically less than or equal to the chosen significance level, α) indicates that the observed data are unlikely to have occurred if the null hypothesis were true. This provides evidence against the null hypothesis, leading us to reject it.
  - A large P-value (greater than α) suggests that the observed data are reasonably likely to occur even if the null hypothesis were true. In this case, we do not have enough evidence to reject the null hypothesis.
---

5. **How do you interpret the P-value in hypothesis testing?**

  Ans. To interpret the P-value in hypothesis testing, we typically compare it to the chosen significance level (α). Here's how the interpretation works:

  1. Choose a Significance Level (α): This is usually set before the test (e.g., α=0.05). It represents the threshold for how unlikely the observed data must be under the null hypothesis for us to reject it.

  2. Calculate the P-value: This is obtained from the statistical test performed on our sample data.

  3. Compare the P-value to α:

    - If the P-value is less than or equal to α (P≤α):

        - We reject the null hypothesis (H0).
        - This indicates that the observed data (or more extreme data) would be unlikely if the null hypothesis were true.
        - We have statistically significant evidence in favor of the alternative hypothesis (Ha).
        - In practical terms, we might conclude that there is a real effect or difference.

    - If the P-value is greater than α (P>α):
        
        - We fail to reject the null hypothesis (H0).
        - This indicates that the observed data are reasonably likely to occur if the null hypothesis were true.
        - We do not have enough statistical evidence to support the alternative hypothesis.
        - In practical terms, we cannot conclude that there is a real effect or difference based on this test. This does not mean we've proven the null hypothesis is true, only that we haven't found sufficient evidence to reject it.

6. **What are Type 1 and Type 2 errors in hypothesis testing?**

  Ans. In hypothesis testing, there are two types of errors we can make when deciding whether to reject or fail to reject the null hypothesis:

  - Type I Error (False Positive):
      - This occurs when we reject the null hypothesis (H0) when it is actually true.
      - We conclude there is an effect or a difference when there isn't one in the population.
      - The probability of making a Type I error is denoted by α (alpha), which is the significance level we set for our test.
      - Think of it as saying something is true when it's not.

  - Type II Error (False Negative):
      - This occurs when we fail to reject the null hypothesis (H0) when it is actually false.
      - We conclude there is no effect or difference when there actually is one in the population.
      - The probability of making a Type II error is denoted by β (beta).
      - The power of a statistical test is the probability of correctly rejecting a false null hypothesis, and it is calculated as 1 - β.
      - Think of it as saying something is not true when it actually is.
---

7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

  Ans. The key difference between a one-tailed and a two-tailed test in hypothesis testing lies in the directionality specified in the alternative hypothesis (Ha).

  1. Two-Tailed Test:

    - The alternative hypothesis (Ha) states that the population parameter is different from the value stated in the null hypothesis (H0). This "different" implies it could be either greater than or less than.
    - Ha : μ != μ0 (where μ is the population mean and μ0 is the value stated in H0)
    - The critical region (the area where you would reject H0) is split into two tails of the sampling distribution, one in each direction (both the lower and upper ends).
    - We are looking for evidence of a difference in either direction.
  - Example of a two-tailed test:
    
    A company wants to know if the average lifespan of their new light bulb is different from 60 days.

    - H0 : μ = 60 days
    - Ha : μ ≠ 60 days

  2. One-Tailed Test:

    - The alternative hypothesis (Ha) specifies that the population parameter is different from the value stated in the null hypothesis (H0) in a particular direction (either greater than OR less than).
    - Right-tailed test: Ha : μ > μ0 (you're only interested if the mean is greater)
    - Left-tailed test: Ha : μ < μ0 (you're only interested if the mean is less)
    - The critical region is located in only one tail of the sampling distribution, corresponding to the direction specified in Ha.
    - We are only looking for evidence of a difference in a specific direction.
  - Examples of one-tailed tests:

    - Right-tailed: A researcher believes that a new fertilizer will increase crop yield.
      - H0 : μ ≤ μ0 (crop yield is less than or equal to the current yield)
      - Ha : μ > μ0 (crop yield is greater than the current yield)
      
    - Left-tailed: A manufacturer believes that a new production process will decrease the defect rate.
      - H0 : p ≥ p0 (defect rate is greater than or equal to the current rate)
      - Ha : p < p0 (defect rate is less than the current rate)

8. **What is the Z-test, and when is it used in hypothesis testing?**

  Ans. The Z-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups, or between a sample mean and a population mean, when the population standard deviation (σ) is known. The test relies on the assumption that the data are normally distributed. For large sample sizes (typically n≥30), the Central Limit Theorem ensures that the sampling distribution of the mean is approximately normal even if the population is not perfectly normal, making the Z-test applicable in more scenarios.

  It is used in hypothesis testing when the following conditions are met:

  - We know the population standard deviation (σ). This is a crucial requirement. If the population standard deviation is unknown, a t-test is usually more appropriate.
  - Our data is approximately normally distributed. This assumption is more critical for small sample sizes. For large sample sizes (n≥30), the Central Limit Theorem can compensate for non-normality.
  - We have a sufficiently large sample size (n≥30). While a Z-test can theoretically be used with smaller samples if the population is normally distributed and σ is known, it's more commonly associated with larger samples due to the reliance on the Central Limit Theorem when normality isn't guaranteed.

9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**

  Ans. We calculate the Z-score using the following formula : Z = (x - μ) / σ

  Where:

  - x is the individual data point.
  - μ (mu) is the population mean.
  - σ (sigma) is the population standard deviation.

  In the context of a Z-test, the calculated Z-score is your test statistic. It represents how many standard deviations the sample mean (x bar) is away from the hypothesized population mean (μ0) under the assumption that the null hypothesis is true.

  - A Z-score close to 0 indicates that our sample mean is close to the hypothesized population mean. This would typically lead to a larger P-value and a failure to reject the null hypothesis.
  - A Z-score far from 0 (either a large positive or a large negative value) indicates that our sample mean is quite different from the hypothesized population mean. This would typically lead to a smaller P-value and a rejection of the null hypothesis (if the P-value is less than the significance level).

  Essentially, the Z-score standardizes the difference between our sample statistic and the null hypothesis value, allowing us to assess how likely such a difference (or a more extreme one) would be if the null hypothesis were true by referring to the standard normal distribution.
---


10. **What is the T-distribution, and when should it be used instead of the normal distribution?**

  Ans. The T-distribution, also known as the Student's t-distribution, is a probability distribution that looks similar to a normal distribution but has heavier tails. This means that extreme values are more likely to occur in a t-distribution compared to a normal distribution. The exact shape of the t-distribution depends on its degrees of freedom, which are typically related to the sample size (often df = n - 1 for a one-sample t-test). As the degrees of freedom increase, the t-distribution approaches the shape of the standard normal distribution.

  We should use the T-distribution instead of the normal distribution primarily when:

  - The population standard deviation (σ) is unknown, and we have to estimate it from the sample standard deviation (s).
  - The sample size (n) is small (generally considered less than 30).

11. **What is the difference between a Z-test and a T-test?**

  Ans.

  - Z-test:

      - Used when we know the standard deviation of the entire population.
      - Also can be used for large sample sizes (n≥30) even if the population standard deviation is unknown because, by the Central Limit Theorem, the sample standard deviation provides a good estimate, and the t-distribution approaches the normal distribution.
      - The test statistic (Z-score) follows a standard normal distribution.

      The formula for a one-sample Z-test is:

      Z = (x bar - μ0) / (σ / √n)

  - T-test:

      - Used when the population standard deviation is unknown, which is more common in practice. In this case, we estimate it using the sample standard deviation (s).
      - Especially appropriate for small sample sizes (n<30) because the t-distribution accounts for the extra uncertainty introduced by estimating the population standard deviation.
      - The test statistic (t-score) follows a t-distribution, which has heavier tails than the normal distribution, especially with fewer degrees of freedom (df=n−1 for a one-sample t-test). As the sample size (and thus degrees of freedom) increases, the t-distribution approaches the standard normal distribution.

      The formula for a one-sample T-test is:

      t =  (x bar - μ0) / (s / √n)

12. **What is the T-test, and how is it used in hypothesis testing?**

  Ans. The T-test is a statistical hypothesis test that is used to determine if there is a significant difference between the means of two groups. It is particularly useful when the population standard deviation is unknown and is often used with smaller sample sizes. The t-test relies on the t-distribution.

  How the T-test is used in hypothesis testing:

 - State the null hypothesis (H0) and the alternative hypothesis (Ha).
      - H0 typically states that there is no difference between the means of the two groups (μ1 = μ2) or that the mean of a single group is equal to some hypothesized value (μ = μ0).
      - Ha states that there is a difference (μ1 ≠ μ2 or μ ≠ μ0), or a directional difference (μ1 > μ2, μ1 < μ2, μ > μ0, or μ < μ0).
  - Choose a significance level (α). This determines the probability of rejecting the null hypothesis when it is true.
  - Collect data and calculate the test statistic (t-score). The formula for the t-score depends on the type of t-test:
      - One-sample t-test: Compares the mean of a single sample to a known or hypothesized population mean.
      
        t = (x bar - μ0) / (s / √n), where: x bar is the sample mean, μ0 is the hypothesized population mean, s is the sample standard deviation, and n is the sample size.
      - Independent two-sample t-test: Compares the means of two independent groups.

        t =  ( (x1 bar - x2 bar) - (μ1 - μ2) ) / √( (s1² / n1) + (s2² / n2) ), where: x1 bar, s1, n1 and x2 bar, s2, n2 are the means, standard deviations, and sample sizes of the two groups, and μ1 - μ2 is the hypothesized difference (often 0).

      - Paired t-test: Compares the means of the same group at two different times or under two different conditions. It looks at the differences within each pair.

        t = d bar / ( sd / √n), where: d bar is the mean of the differences, sd is the standard deviation of the differences, and n is the number of pairs.

  - Determine the degrees of freedom (df). The degrees of freedom are related to the sample size(s) and affect the shape of the t-distribution. For example, for a one-sample t-test, df = n - 1. For an independent two-sample t-test, it's approximately n1 + n2 - 2. For a paired t-test, df = n - 1 (where n is the number of pairs).

  - Find the P-value or the critical value.

      - P-value approach: Find the probability of observing a t-score as extreme as, or more extreme than, the one calculated, under the assumption that the null hypothesis is true. This is done using the t-distribution with the appropriate degrees of freedom.
      - Critical value approach: Find the critical t-value(s) from a t-table based on the significance level (α) and the degrees of freedom.

  - Make a decision.

      - P-value approach: If the P-value is less than or equal to α, reject the null hypothesis.
      - Critical value approach: If the absolute value of the calculated t-score is greater than the critical t-value, reject the null hypothesis.
  - Draw a conclusion in the context of the problem.

13. **What is the relationship between Z-test and T-test in hypothesis testing?**

  Ans. The Z-test and the T-test are both used in hypothesis testing to compare means, but they are appropriate under different conditions. Here's their relationship:

  - Goal: Both tests aim to determine if there is a statistically significant difference between a sample mean (or means) and a population mean (or between two sample means).

  - Test Statistic: Both tests calculate a test statistic that measures how far the sample data deviates from the null hypothesis.

      - Z-test statistic (Z-score):

        Z =  (x bar - μ0) / (σ / √n)

      - T-test statistic (t-score):
      
        t = (x bar - μ0) / (s / √n)

  - Underlying Distribution: The test statistic is evaluated against a different probability distribution:

      - The Z-score is compared to the standard normal distribution.
      - The t-score is compared to the t-distribution, whose shape depends on the degrees of freedom (related to the sample size).

  - Conditions for Use:

      - Z-test: Primarily used when the population standard deviation (σ) is known. It's also often used as an approximation for large sample sizes (n≥30) even if σ is unknown, due to the Central Limit Theorem.
      - T-test: Used when the population standard deviation (σ) is unknown and is estimated by the sample standard deviation (s). It is particularly important for small sample sizes (n<30), but is valid for all sample sizes. As the sample size increases, the t-distribution approaches the standard normal distribution, and the results of a T-test become very similar to those of a Z-test.

14. **What is a confidence interval, and how is it used to interpret statistical results?**

  Ans. A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true value of an unknown population parameter. Instead of providing a single point estimate, a confidence interval gives a lower and upper bound, indicating a range within which the population parameter is expected to lie with a certain level of confidence.

  Key components of a confidence interval:

  - Point Estimate: This is the sample statistic (e.g., sample mean, sample proportion) used to estimate the population parameter. It's usually the midpoint of the confidence interval.
  - Margin of Error: This is the range of values above and below the point estimate that the population parameter is likely to fall within. It depends on the desired confidence level, the variability in the sample, and the sample size.
  - Confidence Level: This is the probability (often expressed as a percentage, like 95%) that if the same population were sampled multiple times and confidence intervals calculated for each sample, the true population parameter would fall within these intervals a certain percentage of the time. For a 95% confidence level, about 95 out of 100 such intervals would contain the true parameter.

  How to interpret statistical results using confidence intervals:

  - Range of Plausible Values: A confidence interval provides a range of values that are plausible for the population parameter. For example, a 95% confidence interval for the average height of adult males might be [5'9", 5'11"]. This suggests that we are 95% confident that the true average height in the population falls within this range.

  - Precision of the Estimate: The width of the confidence interval indicates the precision of the estimate. A narrower interval suggests a more precise estimate of the population parameter, while a wider interval indicates less precision. Precision is often influenced by the sample size and the variability of the data.

  - Statistical Significance (in relation to a null hypothesis): Confidence intervals can be used to assess statistical significance in relation to a hypothesized value (often the value stated in the null hypothesis).

      - If the hypothesized value falls within the confidence interval, then the sample data do not provide enough evidence to reject the null hypothesis at the corresponding significance level (1 - confidence level).
      - If the hypothesized value falls outside the confidence interval, then the sample data provide statistically significant evidence to reject the null hypothesis at that significance level.

15. **What is the margin of error, and how does it affect the confidence interval?**

  Ans. The margin of error is the range of values above and below the sample statistic within which the true population parameter is likely to fall. It quantifies the uncertainty associated with using a sample to estimate a characteristic of the entire population.

  **How the margin of error affects the confidence interval:**

  The margin of error is directly used to construct the confidence interval. The formula for a confidence interval is:

  Confidence Interval=Sample Statistic ± Margin of Error

  This means:

  - The lower bound of the confidence interval is the sample statistic minus the margin of error.
  - The upper bound of the confidence interval is the sample statistic plus the margin of error.

  Therefore, the margin of error determines the width of the confidence interval.

  - A larger margin of error results in a wider confidence interval. This indicates a less precise estimate of the population parameter. A wider interval suggests that the true population parameter could lie within a broader range of values. This typically happens when you have a smaller sample size, higher variability in your data, or a higher desired confidence level.

  - A smaller margin of error results in a narrower confidence interval. This indicates a more precise estimate of the population parameter. A narrower interval suggests that the true population parameter is likely to lie within a smaller range of values. This typically happens with a larger sample size, lower variability in your data, or a lower desired confidence level.

16. **How is Bayes' Theorem used in statistics, and what is its significance?**

  Ans. Bayes' Theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of a hypothesis based on new evidence. It allows us to calculate the posterior probability of an event (the probability after observing some data) given its prior probability (the probability before observing any data) and the likelihood of the evidence.

  The formula for Bayes' Theorem is: P(A | B) = (P(B | A) . P(A)) / P(B)

  Where:

  - P(A | B) is the posterior probability of event A occurring given that event B has occurred.
  - P(B | A) is the likelihood of event B occurring given that event A has occurred.
  - P(A) is the prior probability of event A occurring.
  - P(B) is the marginal likelihood or the probability of event B occurring. It can also be expanded using the law of total probability if needed.

  How it's used in statistics:

  - Updating Beliefs: The core of Bayes' Theorem is its ability to update our beliefs (represented as probabilities) in light of new data. We start with a prior belief P(A), observe evidence B, and then use Bayes' Theorem to calculate the updated belief P(A | B).

  - Bayesian Inference: Bayes' Theorem is the cornerstone of Bayesian statistics, a major approach to statistical inference. In Bayesian inference:

      - We start with a prior distribution over the parameters we are interested in.
      - We observe data, which gives us the likelihood of the data given the parameters.
      - We use Bayes' Theorem to combine the prior and the likelihood to obtain a posterior distribution over the parameters. This posterior distribution represents our updated beliefs about the parameters after seeing the data.

  - Classification Problems: Bayes' Theorem is used in various classification algorithms, such as Naive Bayes classifiers, which are popular in machine learning for tasks like spam detection, text classification, and medical diagnosis.

  - Medical Diagnosis: It can be used to calculate the probability of a disease given the results of a diagnostic test, taking into account the prevalence of the disease and the accuracy of the test (sensitivity and specificity).

  Significance of Bayes' Theorem:

  - Principled way to incorporate prior knowledge: Unlike frequentist statistics, Bayesian methods explicitly allow the incorporation of prior beliefs or knowledge into the analysis. This can be crucial when data is scarce or when there is existing expertise.

  - Provides a probability distribution of parameters: Bayesian inference yields a probability distribution (the posterior) for the parameters of interest, rather than just a point estimate. This allows for a more nuanced understanding of the uncertainty associated with the estimates.

  - Foundation for Bayesian Statistics: Bayes' Theorem is the central pillar of Bayesian statistics, which offers a different paradigm for statistical inference compared to frequentist statistics. Bayesian methods are increasingly used in various fields due to their flexibility and interpretability.

  - Intuitive interpretation: The idea of updating beliefs based on evidence resonates intuitively with how we often learn and make decisions in everyday life.

17. **What is the Chi-square distribution, and when is it used?**

  Ans. The Chi-square (χ2) distribution is a continuous probability distribution that arises frequently in statistics, particularly in hypothesis testing. It is the distribution of the sum of the squares of k independent standard normal random variables, where k is the degrees of freedom.

  Some common applications of Chi-square distribution are:

  - Goodness-of-fit tests: To determine if the observed frequencies of categories in a sample match an expected distribution. For example, testing if the number of heads and tails in a series of coin flips matches a 50/50 expectation.

  - Tests of independence (or association): To examine whether there is a statistically significant association between two or more categorical variables. This is often done using a contingency table. For example, testing if there is a relationship between gender and preference for a certain product.

  - Tests of homogeneity: To determine if the distribution of a categorical variable is the same across different populations or groups. For example, comparing the distribution of political affiliations across different age groups.

  - Hypothesis testing about the variance of a normally distributed population: The Chi-square distribution can be used to construct confidence intervals and perform hypothesis tests for a single population variance.

18. **What is the Chi-square goodness of fit test, and how is it applied?**

  Ans. The Chi-square goodness of fit test is a statistical hypothesis test used to determine if the observed frequencies of a categorical variable match an expected distribution. In simpler terms, it assesses how well our observed data "fits" a theoretical distribution we might expect (e.g., a uniform distribution where all categories are equally likely, or a distribution based on prior knowledge).

  How it is applied:

  - State the null hypothesis (H0) and the alternative hypothesis (Ha).
      - H0 : The observed frequencies follow the specified distribution.
      - Ha : The observed frequencies do not follow the specified distribution.

  - Choose a significance level (α). This is the probability of rejecting the null hypothesis when it is true.

  - Collect observed data. Count the number of observations in each category of our categorical variable.

  - Determine the expected frequencies. Based on our hypothesized distribution, calculate how many observations we would expect in each category if the null hypothesis were true.

  - Calculate the Chi-square test statistic (χ2). The formula is:
      
      χ2 = ∑( (Oi - Ei)² / Ei)

      where:

      Oi is the observed frequency in category i.
      
      Ei is the expected frequency in category i.

      The summation (∑) is over all categories.

  - Determine the degrees of freedom (df). For a goodness of fit test, the degrees of freedom are typically df = k - 1, where k is the number of categories. If we estimate parameters from the data to determine the expected frequencies, we would subtract an additional degree of freedom for each estimated parameter.


  - Find the P-value or the critical value.

      - P-value approach: Find the probability of obtaining a Chi-square statistic as extreme as, or more extreme than, the one calculated, given that the null hypothesis is true. This is done using the Chi-square distribution with the appropriate degrees of freedom.
      - Critical value approach: Find the critical Chi-square value from a Chi-square distribution table based on the chosen significance level (α) and the degrees of freedom.

  - Make a decision.

      - P-value approach: If the P-value is less than or equal to α, reject the null hypothesis.
      - Critical value approach: If the calculated Chi-square statistic is greater than the critical value, reject the null hypothesis.

  - Draw a conclusion. Based on our decision, conclude whether there is enough evidence to say that the observed distribution differs significantly from the expected distribution.

19. **What is the F-distribution, and when is it used in hypothesis testing?**

  Ans. The F-distribution is a continuous probability distribution that arises as the ratio of two independent chi-square distributed random variables, each divided by its degrees of freedom. It is named after Sir Ronald Fisher.

  Key characteristics of the F-distribution:

  - It is continuous and only takes non-negative values.
  - Its shape is determined by two sets of degrees of freedom: one for the numerator (df1) and one for the denominator (df2).
  - It is skewed to the right. The skewness decreases as the degrees of freedom increase.

  The F-distribution is primarily used in F-tests, which are employed in several statistical contexts, most notably:

  - Analysis of Variance (ANOVA): To test for the equality of the means of two or more populations. In ANOVA, the F-statistic is the ratio of the variance between groups to the variance within groups. A significant F-statistic suggests that at least one of the group means is different from the others.

  - Comparing Variances of Two Populations: To test if the variances of two normally distributed populations are equal. The F-statistic in this case is the ratio of the two sample variances.

  - Testing the Overall Significance of a Regression Model: In regression analysis, the F-test can be used to determine if the model as a whole is a better fit to the data than a model with no predictors.

  - Comparing Nested Regression Models: To determine if adding a set of predictors significantly improves the fit of a regression model.

20. **What is an ANOVA test, and what are its assumptions?**

  Ans. The ANOVA (Analysis of Variance) test is a statistical method used to compare the means of two or more groups. Unlike t-tests, which are typically used to compare the means of only two groups, ANOVA allows for the simultaneous comparison of multiple group means. It does this by analyzing the variance within each group compared to the variance between the groups.

  The fundamental idea behind ANOVA is to see if the differences between the group means are large enough to be statistically significant, rather than just due to random chance.

  There are different types of ANOVA, including:

  - One-way ANOVA: Used when we have one independent categorical variable and we want to see if it has an effect on a continuous dependent variable across three or more groups.
  - Two-way ANOVA: Used when we have two independent categorical variables and want to examine their individual and combined effects on a continuous dependent variable.
  - Repeated Measures ANOVA: Used when we are measuring the same subjects multiple times under different conditions.

  Assumptions of ANOVA:

  For the results of an ANOVA test to be valid, several assumptions should be met:

  - Independence of Observations: The data points within each group and between groups should be independent of each other. This means that one observation should not influence another. This is often related to the study design and how data was collected.

  - Normality: The dependent variable should be approximately normally distributed for each group. ANOVA is somewhat robust to minor violations of this assumption, especially with larger sample sizes. However, significant deviations from normality can affect the validity of the results. We can check this by looking at histograms, Q-Q plots, or using normality tests (like Shapiro-Wilk).

  - Homogeneity of Variances (Homoscedasticity): The variances of the dependent variable should be approximately equal across all groups. If the variances are very different (heteroscedasticity), it can lead to inaccurate F-statistic and p-values. We can test this assumption using tests like Levene's test or Bartlett's test. If this assumption is violated, adjustments to the ANOVA or alternative tests (like Welch's ANOVA) might be necessary.

21. **What are the different types of ANOVA tests?**

  Ans. There are several types of ANOVA tests, each designed for different research scenarios based on the number of independent variables and how the data is structured. Here are the main types:

  - One-Way ANOVA:

      - Used when we have one independent categorical variable (also called a factor) with three or more levels (groups) and one continuous dependent variable.
      - The goal is to determine if there are any statistically significant differences between the means of these groups.
      - Example: Comparing the test scores of students using three different teaching methods (lecture, workshop, online learning).
      - Assumptions:
        - Independence of observations.
        - Normality of the dependent variable within each group.
        - Homogeneity of variances (equal variances across groups).

  - Two-Way ANOVA:

      - Used when we have two independent categorical variables (factors), each with two or more levels, and one continuous dependent variable.
      - It examines the main effect of each independent variable on the dependent variable, as well as the interaction effect between the two independent variables (whether the effect of one independent variable depends on the level of the other).
      - Example: Examining the effect of both teaching method (lecture, online) and study time (low, high) on student exam scores.
      - Assumptions:
        - Independence of observations.
        - Normality of the dependent variable for each combination of groups of the two independent variables.
        - Homogeneity of variances across all combinations of groups.

  - Repeated Measures ANOVA:

      - Used when we are measuring the same subjects multiple times under different conditions or at different time points. This means the observations are not independent.
      - It examines the effect of one or more within-subject independent variables on a continuous dependent variable.
      - Example: Measuring a patient's pain level at three different time points after a treatment.
      - Assumptions:
        - The dependent variable should be approximately normally distributed at each time point or condition.
        - Sphericity: The variances of the differences between all pairs of related groups (time points or conditions) should be equal. This is a key assumption for repeated measures ANOVA.
---

22. **What is the F-test, and how does it relate to hypothesis testing?**

  Ans. The F-test is a statistical test that uses the F-distribution. It's often used in hypothesis testing to assess the equality of variances between two or more populations or to test the overall significance of a statistical model.

  How the F-test relates to hypothesis testing:

  - Formulate Hypotheses: Like other hypothesis tests, we start by stating a null hypothesis (H0) and an alternative hypothesis (Ha). The exact form of these hypotheses depends on what we are testing (e.g., equality of variances, equality of means in ANOVA, significance of a regression model).

  - Calculate the F-statistic: The F-test calculates a test statistic called the F-statistic. The formula for the F-statistic varies depending on the specific application of the F-test. For example, when comparing the variances of two samples, the F-statistic is the ratio of the two sample variances:
  
      F = (s1² - s2²)

      In ANOVA, the F-statistic is the ratio of the variance between groups to the variance within groups. In regression, it compares the variance explained by the model to the unexplained variance.

  - Determine the Degrees of Freedom: The F-distribution is characterized by two sets of degrees of freedom: the numerator degrees of freedom (df1) and the denominator degrees of freedom (df2). These are determined by the sample sizes and the structure of the data, depending on the test being performed.

  - Find the P-value or Critical Value:
      - We compare the calculated F-statistic to the F-distribution with the appropriate degrees of freedom.
      - The p-value is the probability of observing an F-statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true.
      - Alternatively, we can compare the F-statistic to a critical value from the F-distribution at our chosen significance level (α).
      
  - Make a Decision:
      - If the p-value is less than or equal to α, we reject the null hypothesis.
      - If the F-statistic is greater than the critical value, we reject the null hypothesis.

  - Draw a Conclusion: Based on our decision, we interpret the results in the context of our research question. For instance, in ANOVA, rejecting the null hypothesis would suggest that there is a statistically significant difference between at least two of the group means.

```python
def bernoulli_pdf(x, p):
  """
  Calculates the Probability Mass Function (PMF) of a Bernoulli distribution.

  The Bernoulli distribution models a single trial of a random experiment
  that has only two possible outcomes: success (usually denoted by 1)
  or failure (usually denoted by 0).

  Args:
    x (int): The outcome (0 for failure, 1 for success).
    p (float): The probability of success (0 <= p <= 1).

  Returns:
    float: The probability of the given outcome. Returns None if x is not 0 or 1
           or if p is not a valid probability.
  """
  if not (0 <= p <= 1):
    print("Error: Probability 'p' must be between 0 and 1.")
    return None
  if x == 1:
    return p
  elif x == 0:
    return 1 - p
  else:
    print("Error: Outcome 'x' must be either 0 or 1 for Bernoulli distribution.")
    return None

if __name__ == '__main__':
  probability_of_success = 0.3

  # Probability of success (x=1)
  prob_success = bernoulli_pdf(1, probability_of_success)
  if prob_success is not None:
    print(f"P(X=1) = {prob_success}")

  # Probability of failure (x=0)
  prob_failure = bernoulli_pdf(0, probability_of_success)
  if prob_failure is not None:
    print(f"P(X=0) = {prob_failure}")

  # Test with an invalid outcome
  invalid_prob = bernoulli_pdf(2, probability_of_success)
  if invalid_prob is not None:
    print(f"P(X=2) = {invalid_prob}")

  # Test with an invalid probability
  invalid_p = bernoulli_pdf(1, 1.5)
  if invalid_p is not None:
    print(f"P(X=1) with p=1.5: {invalid_p}")
```

6. **Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(low, high, values):
  """
  Calculates the Cumulative Distribution Function (CDF) for given values
  of a discrete uniform distribution.

  Args:
    low (int): The lower bound of the distribution (inclusive).
    high (int): The upper bound of the distribution (inclusive).
    values (list or numpy.ndarray): The values at which to evaluate the CDF.

  Returns:
    numpy.ndarray: An array of CDF values corresponding to the input values.
                   Returns None if low > high.
  """
  if low > high:
    print("Error: Lower bound cannot be greater than the upper bound.")
    return None

  cdf_values = []
  num_possible = high - low + 1
  if num_possible <= 0:
      return np.array([0.0] * len(values)) # Handle case where low == high

  for x in values:
    if x < low:
      cdf = 0.0
    elif x > high:
      cdf = 1.0
    else:
      cdf = (x - low + 1) / num_possible
    cdf_values.append(cdf)
  return np.array(cdf_values)

if __name__ == "__main__":
  lower_bound = 1
  upper_bound = 5

  # Generate values to evaluate the CDF at
  x_values = np.linspace(lower_bound - 1, upper_bound + 1, 100)

  # Calculate the CDF
  cdf = discrete_uniform_cdf(lower_bound, upper_bound, x_values)

  if cdf is not None:
    # Plot the CDF
    plt.figure(figsize=(8, 6))
    plt.step(x_values, cdf, where='post', marker='o', linestyle='-', color='blue')
    plt.xlabel("Value (x)")
    plt.ylabel("CDF(x)")
    plt.title(f"Cumulative Distribution Function (CDF) of Discrete Uniform Distribution [{lower_bound}, {upper_bound}]")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.yticks(np.arange(0, 1.1, 0.1))
    plt.xticks(np.arange(lower_bound - 1, upper_bound + 2, 1))
    plt.ylim(-0.05, 1.05)
    plt.show()
```

9. **Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # For potentially nicer plots

def calculate_z_scores(data):
  """
  Calculates the Z-scores of a dataset.

  Args:
    data (list or numpy.ndarray): The input dataset.

  Returns:
    numpy.ndarray: An array of Z-scores. Returns None if the input data
                   is empty or has a standard deviation of zero.
  """
  data_array = np.array(data)
  if data_array.size == 0:
    print("Error: Input data is empty.")
    return None

  mean = np.mean(data_array)
  std_dev = np.std(data_array)

  if std_dev == 0:
    print("Warning: Standard deviation is zero. All Z-scores will be undefined (or zero if treated as such).")
    return np.zeros_like(data_array, dtype=float)  # Return zeros in this case

  z_scores = (data_array - mean) / std_dev
  return z_scores

def plot_z_scores(z_scores, original_data=None):
  """
  Plots the Z-scores. Optionally plots the original data as well.

  Args:
    z_scores (numpy.ndarray): The array of Z-scores.
    original_data (list or numpy.ndarray, optional): The original dataset.
                                                     Defaults to None.
  """
  if z_scores is None:
    print("No Z-scores to plot.")
    return

  n = len(z_scores)
  indices = np.arange(n)

  plt.figure(figsize=(10, 6))

  if original_data is not None:
    plt.subplot(2, 1, 1)
    plt.plot(indices, original_data, marker='o', linestyle='-', color='skyblue')
    plt.title('Original Data')
    plt.xlabel('Data Point Index')
    plt.ylabel('Value')
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.subplot(2, 1, 2)
    plt.plot(indices, z_scores, marker='o', linestyle='-', color='salmon')
    plt.title('Z-scores')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-score')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # Mark the zero Z-score line
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout() # Adjust subplot parameters for a tight layout.
  else:
    plt.plot(indices, z_scores, marker='o', linestyle='-', color='salmon')
    plt.title('Z-scores')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-score')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # Mark the zero Z-score line
    plt.grid(True, linestyle='--', alpha=0.6)

  plt.show()

if __name__ == "__main__":
  # Example usage
  data = [2, 4, 4, 4, 5, 5, 7, 9]
  print(f"Original data: {data}")

  z_scores = calculate_z_scores(data)
  if z_scores is not None:
    print(f"Z-scores: {z_scores}")
    plot_z_scores(z_scores, original_data=data)

  print("\nExample with identical data points:")
  identical_data = [5, 5, 5, 5, 5]
  z_scores_identical = calculate_z_scores(identical_data)
  if z_scores_identical is not None:
    print(f"Z-scores for identical data: {z_scores_identical}")
    plot_z_scores(z_scores_identical, original_data=identical_data)

  print("\nExample with empty data:")
  empty_data = []
  z_scores_empty = calculate_z_scores(empty_data)
  if z_scores_empty is not None:
    plot_z_scores(z_scores_empty)
```

10. **Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def generate_exponential_data(lam, size):
  """Generates data from an exponential distribution."""
  return np.random.exponential(scale=1/lam, size=size)

def apply_clt(original_data, num_samples, sample_size):
  """
  Applies the Central Limit Theorem by repeatedly sampling from the
  original data and calculating the means of these samples.

  Args:
    original_data (numpy.ndarray): The original non-normally distributed dataset.
    num_samples (int): The number of sample means to calculate.
    sample_size (int): The size of each random sample drawn from the original data.

  Returns:
    numpy.ndarray: An array of sample means.
  """
  sample_means = []
  n = len(original_data)
  for _ in range(num_samples):
    # Draw a random sample with replacement
    random_sample = np.random.choice(original_data, size=sample_size, replace=True)
    sample_mean = np.mean(random_sample)
    sample_means.append(sample_mean)
  return np.array(sample_means)

if __name__ == "__main__":
  # 1. Define the non-normal distribution
  lambda_param = 0.5  # Rate parameter for exponential distribution
  population_size = 10000
  original_population = generate_exponential_data(lambda_param, population_size)

  # Plot the original distribution
  plt.figure(figsize=(12, 8))
  plt.subplot(2, 1, 1)
  sns.histplot(original_population, kde=True, color="skyblue")
  plt.title(f'Original Exponential Distribution (λ={lambda_param})')
  plt.xlabel('Value')
  plt.ylabel('Frequency')

  # 2. Apply CLT by taking many samples and their means
  num_sampling_distributions = 500
  size_of_each_sample = 30  # Experiment with different sample sizes

  sample_means = apply_clt(original_population, num_sampling_distributions, size_of_each_sample)

  # 3. Plot the distribution of the sample means
  plt.subplot(2, 1, 2)
  sns.histplot(sample_means, kde=True, color="lightcoral")
  plt.title(f'Sampling Distribution of Means (Sample Size = {size_of_each_sample}, {num_sampling_distributions} samples)')
  plt.xlabel('Sample Mean')
  plt.ylabel('Frequency')

  plt.tight_layout()
  plt.show()

  # Print some statistics
  print("\n--- Statistics ---")
  print(f"Mean of the original distribution: {np.mean(original_population):.4f}")
  print(f"Standard deviation of the original distribution: {np.std(original_population):.4f}")
  print(f"Mean of the sample means: {np.mean(sample_means):.4f}")
  print(f"Standard deviation of the sample means (approx. SE): {np.std(sample_means):.4f}")
  print(f"Theoretical Standard Error (σ/√n): {np.std(original_population) / np.sqrt(size_of_each_sample):.4f}")
```

13. **Generate random variables and calculate their corresponding probabilities using the binomial distribution.**

```python
import numpy as np
from scipy.stats import binom

def binomial_probabilities(n, p, num_random_variables=5):
  """
  Generates random variables from a binomial distribution and calculates
  their corresponding probabilities (PMF).

  Args:
    n (int): The number of trials.
    p (float): The probability of success in each trial (0 <= p <= 1).
    num_random_variables (int): The number of random variables to generate
                                 and their probabilities to calculate.

  Returns:
    dict: A dictionary where keys are the generated random variables (number
          of successes) and values are their corresponding probabilities.
          Returns None if p is not a valid probability.
  """
  if not (0 <= p <= 1):
    print("Error: Probability 'p' must be between 0 and 1.")
    return None

  random_variables = np.random.binomial(n, p, size=num_random_variables)
  probabilities = {}
  for k in random_variables:
    # Calculate the Probability Mass Function (PMF) for k successes
    probability = binom.pmf(k, n, p)
    probabilities[k] = probability

  return probabilities

if __name__ == "__main__":
  num_trials = 10
  probability_of_success = 0.4
  num_rv = 7

  results = binomial_probabilities(num_trials, probability_of_success, num_rv)

  if results:
    print(f"Binomial Distribution (n={num_trials}, p={probability_of_success})")
    print("Generated Random Variables (Number of Successes) and their Probabilities:")
    for rv, prob in results.items():
      print(f"P(X = {rv}) = {prob:.4f}")

  print("\n--- Example with invalid probability ---")
  invalid_results = binomial_probabilities(5, 1.2, 3)
  if invalid_results is None:
    print("Function handled invalid probability correctly.")
```

15. **Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy import stats

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05, alternative='two-sided'):
  """
  Performs a Z-test for the mean of a sample.

  Args:
    sample_mean (float): The mean of the sample.
    population_mean (float): The hypothesized mean of the population.
    population_std (float): The standard deviation of the population.
    sample_size (int): The number of observations in the sample.
    alpha (float, optional): The significance level. Defaults to 0.05.
    alternative (str, optional): The alternative hypothesis ('two-sided', 'less', 'greater').
                                 Defaults to 'two-sided'.

  Returns:
    tuple: (z_statistic, p_value, reject_null)
           - z_statistic: The calculated Z-statistic.
           - p_value: The p-value of the test.
           - reject_null (bool): True if the null hypothesis is rejected, False otherwise.
  """
  if population_std <= 0:
    raise ValueError("Population standard deviation must be positive.")

  # Calculate the Z-statistic
  z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

  # Calculate the p-value
  if alternative == 'two-sided':
    p_value = 2 * stats.norm.cdf(-np.abs(z_statistic))
  elif alternative == 'less':
    p_value = stats.norm.cdf(z_statistic)
  elif alternative == 'greater':
    p_value = 1 - stats.norm.cdf(z_statistic)
  else:
    raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

  # Determine whether to reject the null hypothesis
  reject_null = p_value < alpha

  return z_statistic, p_value, reject_null

if __name__ == "__main__":
  # Example scenario:
  # A manufacturer claims that the average lifespan of their light bulbs is 1000 hours
  # with a population standard deviation of 100 hours. A consumer group tests a
  # sample of 40 light bulbs and finds the sample mean lifespan to be 975 hours.
  # We want to test if there is significant evidence to reject the manufacturer's claim
  # at a significance level of 0.05.

  sample_mean_lifespan = 975
  hypothesized_population_mean = 1000
  population_std_dev = 100
  sample_size_tested = 40
  significance_level = 0.05

  print("--- One-Sample Z-Test ---")
  print(f"Sample Mean: {sample_mean_lifespan}")
  print(f"Hypothesized Population Mean: {hypothesized_population_mean}")
  print(f"Population Standard Deviation: {population_std_dev}")
  print(f"Sample Size: {sample_size_tested}")
  print(f"Significance Level (alpha): {significance_level}")

  # Two-sided test: H0: mu = 1000, H1: mu != 1000
  z_stat, p_val, reject = z_test(sample_mean_lifespan, hypothesized_population_mean,
                                  population_std_dev, sample_size_tested,
                                  alpha=significance_level, alternative='two-sided')
  print("\nTwo-Sided Test:")
  print(f"Z-statistic: {z_stat:.4f}")
  print(f"P-value: {p_val:.4f}")
  if reject:
    print(f"P-value ({p_val:.4f}) < alpha ({significance_level}). Reject the null hypothesis.")
    print("There is significant evidence that the average lifespan is different from 1000 hours.")
  else:
    print(f"P-value ({p_val:.4f}) >= alpha ({significance_level}). Fail to reject the null hypothesis.")
    print("There is not enough significant evidence to say the average lifespan is different from 1000 hours.")

  # One-sided test (less): H0: mu >= 1000, H1: mu < 1000
  z_stat_less, p_val_less, reject_less = z_test(sample_mean_lifespan, hypothesized_population_mean,
                                               population_std_dev, sample_size_tested,
                                               alpha=significance_level, alternative='less')
  print("\nOne-Sided Test (less than):")
  print(f"Z-statistic: {z_stat_less:.4f}")
  print(f"P-value: {p_val_less:.4f}")
  if reject_less:
    print(f"P-value ({p_val_less:.4f}) < alpha ({significance_level}). Reject the null hypothesis.")
    print("There is significant evidence that the average lifespan is less than 1000 hours.")
  else:
    print(f"P-value ({p_val_less:.4f}) >= alpha ({significance_level}). Fail to reject the null hypothesis.")
    print("There is not enough significant evidence to say the average lifespan is less than 1000 hours.")

  # One-sided test (greater): H0: mu <= 1000, H1: mu > 1000
  z_stat_greater, p_val_greater, reject_greater = z_test(sample_mean_lifespan, hypothesized_population_mean,
                                                     population_std_dev, sample_size_tested,
                                                     alpha=significance_level, alternative='greater')
  print("\nOne-Sided Test (greater than):")
  print(f"Z-statistic: {z_stat_greater:.4f}")
  print(f"P-value: {p_val_greater:.4f}")
  if reject_greater:
    print(f"P-value ({p_val_greater:.4f}) < alpha ({significance_level}). Reject the null hypothesis.")
    print("There is significant evidence that the average lifespan is greater than 1000 hours.")
  else:
    print(f"P-value ({p_val_greater:.4f}) >= alpha ({significance_level}). Fail to reject the null hypothesis.")
    print("There is not enough significant evidence to say the average lifespan is greater than 1000 hours.")
```

16. **Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence=0.95):
  """
  Calculates the confidence interval for the mean of a dataset.

  Args:
    data (list or numpy.ndarray): The sample dataset.
    confidence (float, optional): The confidence level (between 0 and 1).
                                   Defaults to 0.95 (95% confidence).

  Returns:
    tuple: A tuple containing the lower and upper bounds of the confidence interval.
           Returns None if the input data is empty.
  """
  data_array = np.array(data)
  n = len(data_array)
  if n == 0:
    print("Error: Input data is empty.")
    return None

  mean = np.mean(data_array)
  std_err = stats.sem(data_array)  # Standard error of the mean

  # For large samples (n >= 30), we can use the Z-distribution approximation.
  # For smaller samples, a t-distribution is more appropriate.
  if n >= 30:
    h = std_err * stats.norm.ppf((1 + confidence) / 2)
  else:
    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

  lower_bound = mean - h
  upper_bound = mean + h

  return lower_bound, upper_bound

if __name__ == "__main__":
  # Example dataset: Lifespans of light bulbs (in hours)
  light_bulb_lifespans = [950, 980, 1010, 1030, 990, 1000, 960, 1020, 970, 1040,
                           985, 1015, 995, 1005, 975, 1035, 965, 1025, 998, 1002,
                           1012, 988, 978, 1022, 1008, 992, 1032, 968, 1018, 982]

  confidence_level = 0.95
  confidence_interval = calculate_confidence_interval(light_bulb_lifespans, confidence_level)

  if confidence_interval:
    lower, upper = confidence_interval
    print(f"Dataset: {light_bulb_lifespans}")
    print(f"\n{confidence_level * 100:.0f}% Confidence Interval for the Mean: ({lower:.2f}, {upper:.2f})")
    print("\nInterpretation:")
    print(f"We are {confidence_level * 100:.0f}% confident that the true average lifespan of these light bulbs lies within the interval ({lower:.2f}, {upper:.2f}) hours.")
```

19. **Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def poisson_cdf(k, lam):
  """
  Calculates the Cumulative Distribution Function (CDF) of a Poisson
  distribution at a given value k.

  Args:
    k (int): The number of events for which to calculate the CDF.
             Must be non-negative.
    lam (float): The average rate of events (lambda). Must be non-negative.

  Returns:
    float: The CDF value at k, which is P(X <= k).
           Returns None if inputs are invalid.
  """
  if lam < 0:
    print("Error: Lambda (average rate) must be non-negative.")
    return None
  if not isinstance(k, int) or k < 0:
    print("Error: k (number of events) must be a non-negative integer.")
    return None

  return poisson.cdf(k, lam)

def plot_poisson_cdf(lam, max_k=15):
  """
  Plots the Cumulative Distribution Function (CDF) of a Poisson distribution.

  Args:
    lam (float): The average rate of events (lambda). Must be non-negative.
    max_k (int, optional): The maximum number of events to plot up to.
                           Defaults to 15.
  """
  if lam < 0:
    print("Error: Lambda (average rate) must be non-negative.")
    return

  k_values = np.arange(0, max_k + 1)
  cdf_values = [poisson.cdf(k, lam) for k in k_values]

  plt.figure(figsize=(10, 6))
  plt.step(k_values, cdf_values, where='post', marker='o', linestyle='-', color='skyblue')
  plt.title(f'Cumulative Distribution Function (CDF) of Poisson(λ={lam})')
  plt.xlabel('Number of Events (k)')
  plt.ylabel('P(X ≤ k)')
  plt.xticks(k_values)
  plt.grid(axis='y', linestyle='--', alpha=0.7)
  plt.ylim(-0.05, 1.05)
  plt.show()

if __name__ == "__main__":
  lambda_value = 3  # Average rate of events

  # Calculate CDF for a specific k
  k_event = 2
  cdf_at_k = poisson_cdf(k_event, lambda_value)
  if cdf_at_k is not None:
    print(f"The probability of observing {k_event} or fewer events when λ={lambda_value} is: {cdf_at_k:.4f}")
    print("\nInterpretation:")
    print(f"There is a {cdf_at_k:.2%} chance of observing at most {k_event} events occurring within the given interval.")

  print("\n--- Plotting the CDF ---")
  plot_poisson_cdf(lambda_value, max_k=10)
```

```python
import numpy as np

def calculate_range_iqr(data):
  """
  Calculates the range and interquartile range (IQR) of a dataset.

  Args:
    data (list or numpy.ndarray): The input dataset.

  Returns:
    tuple: A tuple containing the range (min, max) and the IQR.
           Returns None if the input data is empty.
  """
  data_array = np.array(data)
  if data_array.size == 0:
    print("Error: Input data is empty.")
    return None

  data_sorted = np.sort(data_array)
  minimum = data_sorted[0]
  maximum = data_sorted[-1]
  data_range = (minimum, maximum)

  q1 = np.percentile(data_sorted, 25)
  q3 = np.percentile(data_sorted, 75)
  iqr = q3 - q1

  return data_range, iqr

if __name__ == "__main__":
  # Generate data from a normal distribution
  np.random.seed(42)  # for reproducibility
  mean = 0
  std_dev = 1
  num_samples = 1000
  normal_data = np.random.normal(mean, std_dev, num_samples)

  # Calculate the range and IQR
  results = calculate_range_iqr(normal_data)

  if results:
    data_range, iqr_value = results
    print("Dataset generated from a normal distribution:")
    print(f"First 10 data points: {normal_data[:10]}")
    print(f"Number of data points: {len(normal_data)}")
    print(f"\nRange (Min, Max): {data_range}")
    print(f"Interquartile Range (IQR): {iqr_value:.4f}")
```

23. **Implement Z-score normalization on a dataset and visualize its transformation.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def z_score_normalize(data):
  """
  Performs Z-score normalization on a dataset.

  Args:
    data (list or numpy.ndarray): The input dataset.

  Returns:
    numpy.ndarray: The Z-score normalized dataset.
                   Returns None if the input data is empty or has zero standard deviation.
  """
  data_array = np.array(data)
  if data_array.size == 0:
    print("Error: Input data is empty.")
    return None

  mean = np.mean(data_array)
  std_dev = np.std(data_array)

  if std_dev == 0:
    print("Warning: Standard deviation is zero. Cannot perform Z-score normalization.")
    return None

  normalized_data = (data_array - mean) / std_dev
  return normalized_data

def visualize_normalization(original_data, normalized_data, title="Z-score Normalization"):
  """
  Visualizes the original and Z-score normalized datasets.

  Args:
    original_data (list or numpy.ndarray): The original dataset.
    normalized_data (numpy.ndarray): The Z-score normalized dataset.
    title (str, optional): The title of the plot. Defaults to "Z-score Normalization".
  """
  plt.figure(figsize=(12, 6))

  plt.subplot(1, 2, 1)
  sns.histplot(original_data, kde=True, color="skyblue")
  plt.title("Original Data")
  plt.xlabel("Value")
  plt.ylabel("Density/Frequency")

  plt.subplot(1, 2, 2)
  sns.histplot(normalized_data, kde=True, color="salmon")
  plt.title("Z-score Normalized Data")
  plt.xlabel("Z-score")
  plt.ylabel("Density/Frequency")

  plt.suptitle(title)
  plt.tight_layout()
  plt.show()

if __name__ == "__main__":
  # Example dataset (can be any distribution)
  np.random.seed(42)
  original_data = np.concatenate([np.random.normal(loc=5, scale=2, size=300),
                                  np.random.exponential(scale=3, size=200)])

  print("Original Data:")
  print(f"  Mean: {np.mean(original_data):.4f}")
  print(f"  Standard Deviation: {np.std(original_data):.4f}")

  # Perform Z-score normalization
  normalized_data = z_score_normalize(original_data)

  if normalized_data is not None:
    print("\nZ-score Normalized Data:")
    print(f"  Mean: {np.mean(normalized_data):.4f}")
    print(f"  Standard Deviation: {np.std(normalized_data):.4f}")

    # Visualize the transformation
    visualize_normalization(original_data, normalized_data)
```

5. **Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_hypothesis_errors(population_mean_null, population_mean_true, population_std, sample_size, alpha=0.05, alternative='two-sided'):
  """
  Visualizes Type I and Type II errors in hypothesis testing for a one-sample Z-test.

  Args:
    population_mean_null (float): The population mean under the null hypothesis (H0).
    population_mean_true (float): The true population mean (under the alternative hypothesis Ha).
    population_std (float): The population standard deviation.
    sample_size (int): The size of the sample.
    alpha (float, optional): The significance level (Type I error rate). Defaults to 0.05.
    alternative (str, optional): The alternative hypothesis type ('two-sided', 'less', 'greater').
                                 Defaults to 'two-sided'.
  """
  if population_std <= 0:
    raise ValueError("Population standard deviation must be positive.")
  if alternative not in ['two-sided', 'less', 'greater']:
    raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

  # Calculate the standard error of the mean
  sem = population_std / np.sqrt(sample_size)

  # Generate x-values for the sampling distribution under H0 and Ha
  x = np.linspace(population_mean_null - 4 * sem, population_mean_true + 4 * sem, 500)
  y_null = stats.norm.pdf(x, population_mean_null, sem)
  y_true = stats.norm.pdf(x, population_mean_true, sem)

  # Calculate critical Z-values and rejection region
  if alternative == 'two-sided':
    z_critical = stats.norm.ppf(1 - alpha / 2)
    lower_critical_value = population_mean_null - z_critical * sem
    upper_critical_value = population_mean_null + z_critical * sem
    reject_null = (x < lower_critical_value) | (x > upper_critical_value)
    type1_area = 2 * stats.norm.cdf(-z_critical)
  elif alternative == 'less':
    z_critical = stats.norm.ppf(alpha)
    critical_value = population_mean_null + z_critical * sem
    reject_null = x < critical_value
    type1_area = stats.norm.cdf(z_critical)
  elif alternative == 'greater':
    z_critical = stats.norm.ppf(1 - alpha)
    critical_value = population_mean_null + z_critical * sem
    reject_null = x > critical_value
    type1_area = 1 - stats.norm.cdf(z_critical)

  # Calculate Type II error (beta)
  if alternative == 'two-sided':
      beta_lower = stats.norm.cdf((population_mean_null + z_critical * sem - population_mean_true) / sem)
      beta_upper = 1 - stats.norm.cdf((population_mean_null - z_critical * sem - population_mean_true) / sem)
      beta = beta_upper - beta_lower
  elif alternative == 'less':
      beta = stats.norm.cdf((population_mean_null + stats.norm.ppf(alpha) * sem - population_mean_true) / sem)
  elif alternative == 'greater':
        beta = 1 - stats.norm.cdf((population_mean_null + stats.norm.ppf(1-alpha) * sem - population_mean_true) / sem)

  # Plotting
  plt.figure(figsize=(10, 6))
  plt.plot(x, y_null, 'b-', label='Sampling Distribution under H0')
  plt.plot(x, y_true, 'g-', label='Sampling Distribution under Ha')

  # Shade the rejection region
  plt.fill_between(x, y_null, where=reject_null, color='red', alpha=0.3, label='Type I Error (α)')
  # Shade the Type II error region
  plt.fill_between(x, y_true, where=(~reject_null), color='yellow', alpha=0.3, label='Type II Error (β)')

  # Plot critical value(s)
  if alternative == 'two-sided':
    plt.axvline(lower_critical_value, color='r', linestyle='--', linewidth=1)
    plt.axvline(upper_critical_value, color='r', linestyle='--', linewidth=1)
  else:
    plt.axvline(critical_value, color='r', linestyle='--', linewidth=1)

  plt.title('Type I and Type II Errors in Hypothesis Testing')
  plt.xlabel('Sample Mean')
  plt.ylabel('Probability Density')
  plt.legend()
  plt.grid(True)
  plt.show()

  print(f"Type I Error (alpha): {type1_area:.4f}")
  print(f"Type II Error (beta): {beta:.4f}")
  print(f"Power (1 - beta): {1 - beta:.4f}")

if __name__ == "__main__":
  # Example usage
  null_mean = 100
  true_mean = 105  # True population mean is different from the null hypothesis
  pop_std = 15
  sample_size = 30
  alpha_level = 0.05
  alternative_type = 'greater' #can be 'two-sided', 'less', or 'greater'

  visualize_hypothesis_errors(null_mean, true_mean, pop_std, sample_size, alpha_level, alternative_type)
```

```python
import numpy as np
from scipy import stats

def independent_t_test(sample1, sample2, alpha=0.05, alternative='two-sided', equal_var=True):
  """
  Performs an independent samples t-test (Welch's t-test or Student's t-test).

  Args:
    sample1 (list or numpy.ndarray): The data for the first sample.
    sample2 (list or numpy.ndarray): The data for the second sample.
    alpha (float, optional): The significance level. Defaults to 0.05.
    alternative (str, optional): The alternative hypothesis ('two-sided', 'less', 'greater').
                                 Defaults to 'two-sided'.
    equal_var (bool, optional): Assume equal variances (for Student's t-test).
                                Defaults to True. If False, performs Welch's t-test.

  Returns:
    tuple: (t_statistic, p_value, degrees_of_freedom, reject_null)
           - t_statistic (float): The calculated t-statistic.
           - p_value (float): The p-value of the test.
           - degrees_of_freedom (float): The degrees of freedom.
           - reject_null (bool): True if the null hypothesis is rejected, False otherwise.
  """
  if alternative not in ['two-sided', 'less', 'greater']:
    raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

  # Perform the independent samples t-test
  t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=equal_var, alternative=alternative)
  degrees_of_freedom = (len(sample1) + len(sample2)) - 2 if equal_var else None #Degrees of freedom is only easily accessible if equal_var=True

  # Determine whether to reject the null hypothesis
  reject_null = p_value < alpha

  return t_statistic, p_value, degrees_of_freedom, reject_null

def interpret_t_test_results(t_statistic, p_value, degrees_of_freedom, reject_null, alpha, alternative, sample1_name="Sample 1", sample2_name="Sample 2"):
    """
    Interprets the results of the independent samples t-test.

    Args:
      t_statistic (float): The calculated t-statistic.
      p_value (float): The p-value of the test.
      degrees_of_freedom (float): The degrees of freedom.
      reject_null (bool): True if the null hypothesis is rejected, False otherwise.
      alpha (float): The significance level.
      alternative (str): The alternative hypothesis ('two-sided', 'less', 'greater').
      sample1_name (str, optional): Name of the first sample. Defaults to "Sample 1".
      sample2_name (str, optional): Name of the second sample. Defaults to "Sample 2".
    """
    print("\n--- Independent Samples T-Test Results ---")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    if degrees_of_freedom is not None:
      print(f"Degrees of Freedom: {degrees_of_freedom:.2f}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Alternative Hypothesis: {alternative}")

    if reject_null:
        print("\nConclusion: Reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is significant evidence that the means of {sample1_name} and {sample2_name} are different.")
        elif alternative == 'less':
            print(f"There is significant evidence that the mean of {sample1_name} is less than the mean of {sample2_name}.")
        elif alternative == 'greater':
            print(f"There is significant evidence that the mean of {sample1_name} is greater than the mean of {sample2_name}.")
    else:
        print("\nConclusion: Fail to reject the null hypothesis.")
        if alternative == 'two-sided':
            print(f"There is not enough evidence to suggest that the means of {sample1_name} and {sample2_name} are different.")
        elif alternative == 'less':
            print(f"There is not enough evidence to suggest that the mean of {sample1_name} is less than the mean of {sample2_name}.")
        elif alternative == 'greater':
            print(f"There is not enough evidence to suggest that the mean of {sample1_name} is greater than the mean of {sample2_name}.")



if __name__ == "__main__":
    # Example scenario:
    # A researcher wants to compare the test scores of students who used two different study methods.
    # Group 1 used Method A, and Group 2 used Method B.

    group1_scores = [85, 88, 92, 90, 87, 91, 89, 93, 86, 94]
    group2_scores = [78, 82, 80, 85, 79, 81, 83, 84, 77, 86]
    alpha_level = 0.05
    alternative_hypothesis = 'greater'  # Test if Method A scores are greater than Method B scores.
    equal_variance = False # Assume unequal variances (Welch's t-test)

    print("--- Independent Samples T-Test ---")
    print(f"Sample 1 (Method A) Scores: {group1_scores}")
    print(f"Sample 2 (Method B) Scores: {group2_scores}")
    print(f"Significance Level (alpha): {alpha_level}")
    print(f"Alternative Hypothesis: {alternative_hypothesis}")
    print(f"Assuming Equal Variances: {equal_variance}")

    t_stat, p_val, df, reject_h0 = independent_t_test(group1_scores, group2_scores,
                                                          alpha=alpha_level,
                                                          alternative=alternative_hypothesis,
                                                          equal_var=equal_variance)
    interpret_t_test_results(t_stat, p_val, df, reject_h0, alpha_level, alternative_hypothesis, "Method A", "Method B")

    # Example 2: Demonstrating the equal_var parameter
    print("\n--- Example 2: Comparing Equal Variance Assumption ---")
    group3_scores = [65, 70, 75, 72, 68, 73, 71, 74, 69, 76]  # Similar mean, different variance to group1
    print("\nComparing Sample 1 (group1_scores) and Sample 3 (group3_scores) ")
    t_stat_equal_var, p_val_equal_var, df_equal_var, reject_equal_var = independent_t_test(group1_scores, group3_scores, alpha=alpha_level, alternative='two-sided', equal_var=True)
    print("\nAssuming equal variances (Student's t-test):")
    interpret_t_test_results(t_stat_equal_var, p_val_equal_var, df_equal_var, reject_equal_var, alpha_level, 'two-sided', "Sample 1", "Sample 3")

    t_stat_unequal_var, p_val_unequal_var, df_unequal_var, reject_unequal_var = independent_t_test(group1_scores, group3_scores, alpha=alpha_level, alternative='two-sided', equal_var=False)
    print("\nAssuming unequal variances (Welch's t-test):")
    interpret_t_test_results(t_stat_unequal_var, p_val_unequal_var, df_unequal_var, reject_unequal_var, alpha_level, 'two-sided', "Sample 1", "Sample 3")
```

8. **Simulate data and perform both Z-test and T-test, then compare the results using Python.**

```python
import numpy as np
from scipy import stats

# Set a random seed for reproducibility
np.random.seed(42)

# Simulate two independent groups of data
n1 = 50
n2 = 60
mean1 = 10
mean2 = 10.5
std_dev_pop = 2  # Assume we know the population standard deviation for the Z-test
sample1 = np.random.normal(mean1, std_dev_pop, n1)
sample2 = np.random.normal(mean2, std_dev_pop, n2)

print("Sample 1:", sample1[:5])
print("Sample 2:", sample2[:5])
print("-" * 30)

# --- Perform Z-test (assuming known population standard deviation) ---
def independent_z_test(sample1, sample2, pop_std1, pop_std2):
    """Performs an independent two-sample Z-test."""
    n1 = len(sample1)
    n2 = len(sample2)
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    numerator = mean1 - mean2
    denominator = np.sqrt((pop_std1**2 / n1) + (pop_std2**2 / n2))
    z_statistic = numerator / denominator
    # Two-tailed p-value
    p_value = 2 * stats.norm.cdf(-np.abs(z_statistic))
    return z_statistic, p_value

z_statistic, z_p_value = independent_z_test(sample1, sample2, std_dev_pop, std_dev_pop)
print("Z-test results:")
print(f"  Z-statistic: {z_statistic:.3f}")
print(f"  P-value: {z_p_value:.3f}")
print("-" * 30)

# --- Perform T-test (assuming unknown population standard deviation) ---
t_statistic, t_p_value = stats.ttest_ind(sample1, sample2)
print("T-test results:")
print(f"  T-statistic: {t_statistic:.3f}")
print(f"  P-value: {t_p_value:.3f}")
print("-" * 30)

# --- Compare the results ---
print("Comparison of results:")
print(f"Z-test P-value: {z_p_value:.3f}")
print(f"T-test P-value: {t_p_value:.3f}")

if z_p_value < 0.05:
    print("Based on the Z-test, there is a statistically significant difference between the means (alpha=0.05).")
else:
    print("Based on the Z-test, there is no statistically significant difference between the means (alpha=0.05).")

if t_p_value < 0.05:
    print("Based on the T-test, there is a statistically significant difference between the means (alpha=0.05).")
else:
    print("Based on the T-test, there is no statistically significant difference between the means (alpha=0.05).")
```

9. **Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy import stats

def confidence_interval_mean(data, confidence=0.95):
    """
    Calculates the confidence interval for the mean of a sample.

    Args:
        data (array-like): A list or numpy array of the sample data.
        confidence (float, optional): The desired confidence level (between 0 and 1).
                                     Defaults to 0.95 (95% confidence).

    Returns:
        tuple: A tuple containing the lower and upper bounds of the confidence interval.
               Returns (None, None) if the input data is empty.
    """
    if not data:
        return None, None

    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean

    # Calculate the margin of error based on the t-distribution
    # Degrees of freedom = n - 1
    alpha = 1.0 - confidence
    h = std_err * stats.t.ppf(1 - alpha / 2, n - 1)

    lower_bound = mean - h
    upper_bound = mean + h

    return lower_bound, upper_bound

# Example usage:
sample_data = [22, 25, 28, 30, 23, 27, 29, 26, 24, 31]
confidence_level = 0.95
lower, upper = confidence_interval_mean(sample_data, confidence_level)

if lower is not None:
    print(f"Sample Mean: {np.mean(sample_data):.2f}")
    print(f"{confidence_level*100:.0f}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
else:
    print("No data provided to calculate the confidence interval.")
```

10. **Write a Python program to calculate the margin of error for a given confidence level using sample data.**

```python
import numpy as np
from scipy import stats

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for the mean of a sample at a given confidence level.

    Args:
        data (array-like): A list or numpy array of the sample data.
        confidence (float, optional): The desired confidence level (between 0 and 1).
                                     Defaults to 0.95 (95% confidence).

    Returns:
        float: The margin of error. Returns None if the input data is empty.
    """
    if not data:
        return None

    n = len(data)
    std_err = stats.sem(data)  # Standard error of the mean

    # Calculate the critical value from the t-distribution
    # Degrees of freedom = n - 1
    alpha = 1.0 - confidence
    critical_value = stats.t.ppf(1 - alpha / 2, n - 1)

    # Margin of error
    margin_of_error = critical_value * std_err

    return margin_of_error

# Example usage:
sample_data = [45, 52, 48, 55, 50, 49, 53, 51]
confidence_level = 0.90
margin = calculate_margin_of_error(sample_data, confidence_level)

if margin is not None:
    sample_mean = np.mean(sample_data)
    print(f"Sample Data: {sample_data}")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Confidence Level: {confidence_level*100:.0f}%")
    print(f"Margin of Error: {margin:.2f}")

    lower_bound = sample_mean - margin
    upper_bound = sample_mean + margin
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
else:
    print("No data provided to calculate the margin of error.")
```

11. **Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the hypotheses (possible values for the probability of heads)
possible_p_heads = np.linspace(0, 1, 101) # From 0 to 1 with 101 points

# 2. Define the prior probability for each hypothesis
# Let's assume a uniform prior: we initially believe all probabilities of heads are equally likely.
prior = np.ones_like(possible_p_heads) / len(possible_p_heads)

def likelihood(data, p_head):
    """Calculates the likelihood of the observed data given a probability of heads."""
    n_heads = sum(data)
    n_tails = len(data) - n_heads
    # Likelihood of observing n_heads in N flips given p_head (Binomial)
    return (p_head ** n_heads) * ((1 - p_head) ** n_tails)

def bayesian_inference(prior, data, possible_p_heads):
    """Performs Bayesian inference to update the probability of hypotheses given data."""
    likelihoods = np.array([likelihood(data, p) for p in possible_p_heads])

    # Calculate the unnormalized posterior
    unnormalized_posterior = likelihoods * prior

    # Normalize the posterior to get a probability distribution
    posterior = unnormalized_posterior / np.sum(unnormalized_posterior)

    return posterior

# 3. Observe some data (coin flips: True=Heads, False=Tails)
observed_data = [True, True, False, True, True, False, True, True, True, False]
n_flips = len(observed_data)
n_heads = sum(observed_data)
print(f"Observed {n_heads} heads in {n_flips} flips.")

# 4. Perform Bayesian inference
posterior_probability = bayesian_inference(prior, observed_data, possible_p_heads)

# 5. Visualize the results
plt.figure(figsize=(10, 6))
plt.plot(possible_p_heads, prior, 'r--', label='Prior')
plt.plot(possible_p_heads, posterior_probability, 'b-', label='Posterior')
plt.xlabel('Probability of Heads (p_head)')
plt.ylabel('Probability Density')
plt.title('Bayesian Inference for Coin Flip')
plt.legend()
plt.grid(True)
plt.show()

# Find the Maximum A Posteriori (MAP) estimate
map_index = np.argmax(posterior_probability)
map_p_head = possible_p_heads[map_index]
print(f"\nMaximum A Posteriori (MAP) estimate for P(Heads): {map_p_head:.3f}")
```

12. **Perform a Chi-square test for independence between two categorical variables in Python.**

```python
import pandas as pd
from scipy.stats import chi2_contingency

def perform_chi2_independence_test(data, col1, col2, alpha=0.05):
    """
    Performs a Chi-square test for independence between two categorical variables.

    Args:
        data (pd.DataFrame): The DataFrame containing the data.
        col1 (str): The name of the first categorical column.
        col2 (str): The name of the second categorical column.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing:
            - chi2 (float): The Chi-square statistic.
            - p (float): The p-value of the test.
            - dof (int): The degrees of freedom.
            - expected (numpy.ndarray): The expected frequencies.
    """
    # Create a contingency table
    contingency_table = pd.crosstab(data[col1], data[col2])
    print("Contingency Table:")
    print(contingency_table)
    print("-" * 30)

    # Perform the Chi-square test of independence
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    print("Chi-square Test Results:")
    print(f"Chi-square statistic: {chi2:.4f}")
    print(f"P-value: {p:.4f}")
    print(f"Degrees of freedom: {dof}")
    print("\nExpected Frequencies:")
    print(expected)
    print("-" * 30)

    if p < alpha:
        print(f"The p-value ({p:.4f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis, suggesting that there is a significant association between '{}' and '{}'.".format(col1, col2))
    else:
        print(f"The p-value ({p:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis, suggesting that there is no significant association between '{}' and '{}'.".format(col1, col2))

    return chi2, p, dof, expected

# Example Usage:
# Create a sample DataFrame with two categorical variables
data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female'],
    'Smoker': ['Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes']
})

# Perform the Chi-square test for independence between 'Gender' and 'Smoker'
chi2, p, dof, expected = perform_chi2_independence_test(data, 'Gender', 'Smoker')
```

13. **Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.**

```python
import pandas as pd
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculates the expected frequencies for a Chi-square test
    based on the observed contingency table.

    Args:
        observed_data (pd.DataFrame): The observed contingency table.
                                       Rows and columns represent the categories
                                       of the two variables.

    Returns:
        pd.DataFrame: A DataFrame containing the expected frequencies.
                      Returns None if the input is not a DataFrame or is empty.
    """
    if not isinstance(observed_data, pd.DataFrame) or observed_data.empty:
        print("Error: Input must be a non-empty Pandas DataFrame.")
        return None

    row_totals = observed_data.sum(axis=1)
    col_totals = observed_data.sum(axis=0)
    grand_total = observed_data.values.sum()

    expected_frequencies = pd.DataFrame(index=observed_data.index, columns=observed_data.columns)

    for row in observed_data.index:
        for col in observed_data.columns:
            expected_frequency = (row_totals[row] * col_totals[col]) / grand_total
            expected_frequencies.loc[row, col] = expected_frequency

    return expected_frequencies

# Example Usage:
# Observed contingency table
observed = pd.DataFrame({
    'Yes': [20, 15],
    'No': [10, 25]
}, index=['Male', 'Female'])

print("Observed Contingency Table:")
print(observed)
print("-" * 30)

expected = calculate_expected_frequencies(observed)

if expected is not None:
    print("Expected Frequencies:")
    print(expected)
```

14. **Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.**

```python
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected=None, alpha=0.05):
    """
    Performs a Chi-square goodness-of-fit test.

    Args:
        observed (array-like): The observed frequencies.
        expected (array-like, optional): The expected frequencies. If None,
                                        assumes a uniform distribution.
                                        Defaults to None.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing:
            - chi2_statistic (float): The Chi-square test statistic.
            - p_value (float): The p-value of the test.
    """
    observed = np.array(observed)

    if expected is None:
        expected = np.array([np.sum(observed) / len(observed)] * len(observed))
    else:
        expected = np.array(expected)
        if len(observed) != len(expected):
            raise ValueError("Observed and expected arrays must have the same length.")
        if np.any(expected <= 0):
            raise ValueError("Expected frequencies must be greater than zero.")

    chi2_statistic, p_value = chisquare(f_obs=observed, f_exp=expected)

    print("Goodness-of-Fit Test Results:")
    print(f"Chi-square statistic: {chi2_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("We reject the null hypothesis. The observed data does not fit the expected distribution.")
    else:
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}).")
        print("We fail to reject the null hypothesis. The observed data is consistent with the expected distribution.")

    return chi2_statistic, p_value

# Example 1: Testing against a uniform distribution
observed_counts_uniform = [25, 30, 22, 28, 25]
print("\n--- Testing against a uniform distribution ---")
chi2_uniform, p_uniform = goodness_of_fit_test(observed_counts_uniform)

# Example 2: Testing against a specific expected distribution
observed_counts_specific = [60, 85, 70, 75, 60]
expected_proportions = [0.20, 0.25, 0.20, 0.20, 0.15]
total_observed = np.sum(observed_counts_specific)
expected_counts_specific = [p * total_observed for p in expected_proportions]
print("\n--- Testing against a specific expected distribution ---")
chi2_specific, p_specific = goodness_of_fit_test(observed_counts_specific, expected_counts_specific)
```
