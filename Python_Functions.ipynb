1. ** What is the difference between a function and a method in Python?**

  Ans. In Python, while both functions and methods are blocks of reusable code, they have a key distinction: their association with objects. Some of the key differences are:
  - Association: Functions are independent, while methods are associated with objects.
  - Calling: Functions are called by name, methods are called on objects
  - "self": Methods have a "self" parameter, functions do not.

            def __init__(self, name):
                self.name = name
            def bark(self):
                print("Woof!")
        my_dog = Dog("Buddy")
        my_dog.bark()   #output Woof!

2. **Explain the concept of function arguments and parameters in Python.**

  Ans.

  **Parameters**:

  Parameters are the variables listed inside the parentheses in a function's definition. They act as placeholders for the values that will be passed to the function when it is called.

  Example: In the function `def add(x, y):`, x and y are parameters.

  **Arguments**:

  Arguments are the actual values that are passed to a function when it is called. They are the "data" that fills the parameter "slots."

  Example: When we call `add(5, 3)`, 5 and 3 are arguments.

6. **Explain the concept of generators in Python and how they are defined.**

  Ans. In Python, generators are a special type of iterator that allow us to create sequences of values on-the-fly, without storing the entire sequence in memory at once. This makes them highly memory-efficient, especially when dealing with large datasets or infinite sequences.

  Defining Generators:

  Generators are defined using one of two primary methods:
  
  1. Generator Functions:
  - Generator functions look like regular functions, but they use the yield keyword instead of return.
  - When a generator function is called, it returns a generator object.
  - Each time yield is encountered, the function's state is saved, and the yielded value is returned.
  - The function resumes execution from where it left off the next time next() is called on the generator.
  - Example:
          def count_up_to(n):
              i = 1
              while i <= n:
                  yield i
                  i += 1

          my_generator = count_up_to(5)

          print(next(my_generator))  # Output: 1
          print(next(my_generator))  # Output: 2

          for num in my_generator:
              print(num)  # Output: 3, 4, 5

  2. Generator Expressions:
  - Generator expressions are similar to list comprehensions, but they use parentheses () instead of square brackets [ ].
  - They create anonymous generator objects.
  - They are more memory efficient than list comprehensions when working with large datasets.
  - Example:
          squares = (x * x for x in range(1, 6))

          for square in squares:
              print(square)  # Output: 1, 4, 9, 16, 25

7. **What are the advantages of using generators over regular functions?**

  Ans. Generators in Python offer several key advantages over regular functions, particularly when dealing with large datasets or complex data processing:
  
  1. Memory Efficiency:
  - Generators: Produce values on demand using the yield keyword, meaning they only store the current value in memory. This is crucial for handling large datasets or infinite sequences, as it avoids loading everything into memory at once.
  - Regular Functions: Typically compute and store all results in memory before returning them. This can lead to memory issues when dealing with large datasets.
  
  2. Lazy Evaluation:
  - Generators: Employ lazy evaluation, meaning they only compute values when they are needed. This can significantly improve performance, especially when you don't need to process all the data at once.
  - Regular Functions: Evaluate all expressions and computations before returning, even if some of the results are not immediately used.
  
  3. Improved Performance:
  - Generators: Can lead to faster execution times in certain scenarios, especially when dealing with large datasets or when you only need to process a subset of the data.
  - Regular Functions: May require more time to compute and store all results, even if they are not all needed.
  
  4. Code Readability and Conciseness:
  - Generators: Can make your code more concise and readable, especially when dealing with complex data processing tasks. Generator expressions, in particular, provide a compact syntax for creating generators.
  - Regular Functions: May require more verbose code to achieve the same functionality, especially when dealing with iterative processes.
  
  5. Handling Infinite Sequences:
  - Generators: Can be used to represent infinite sequences, as they only generate values when requested. This is not possible with regular functions, which would require infinite memory to store an infinite sequence.
  - Regular Functions: Are limited to finite sequences or data structures that can fit in memory.
  
  6. Iteration Control:
  - Generators: Provide more control over the iteration process, allowing you to pause and resume execution as needed.
  - Regular Functions: Execute to completion in a single pass.

  Example Illustrating Memory Efficiency:
          # Regular function (inefficient for large datasets)
          def square_numbers_list(nums):
              result = []
              for num in nums:
                  result.append(num * num)
              return result

          # Generator function (memory-efficient)
          def square_numbers_generator(nums):
              for num in nums:
                  yield num * num

          large_list = range(1000000) #a very large list.

          #using the list version, will take a long time, and a lot of memory.
          #squares_list = square_numbers_list(large_list)

          #using the generator, is very quick, and uses very little memory.
          squares_generator = square_numbers_generator(large_list)

          #we can then iterate over the generator as needed.
          for i in range(10):
              print(next(squares_generator))

8. **What is a lambda function in Python and when is it typically used?**

  Ans. A lambda function in Python is a small, anonymous function. "Anonymous" means it doesn't have a name like regular functions defined with def. It's defined using the lambda keyword.

  Typical Use Cases:

  Lambda functions are commonly used in situations where you need a small, throwaway function, particularly:
  
  - With `map()`, `filter()`, and `reduce()`:
    - These built-in functions take a function as an argument. Lambda functions are perfect for providing simple, inline functions for these operations.
    - Example:
            numbers = [1, 2, 3, 4, 5]
            squared = list(map(lambda x: x**2, numbers))  # Squaring each number
            even = list(filter(lambda x: x % 2 == 0, numbers)) #filtering for even numbers.
            print(squared) #output [1, 4, 9, 16, 25]
            print(even) #output [2, 4]
  - As keys in `sorted()`:
    - We can use lambda functions to define custom sorting criteria.
    - Example:
            data = [("apple", 3), ("banana", 1), ("cherry", 2)]
            sorted_data = sorted(data, key=lambda item: item[1])  # Sort by the second element
            print(sorted_data) #output [('banana', 1), ('cherry', 2), ('apple', 3)]
  - In GUI programming (event handlers):
    - For simple event handlers, lambda functions can be used to define short, inline functions.
  - For creating simple, one-time functions:
    - When we need a function for a very specific task that won't be reused elsewhere, a lambda function can be more concise than defining a full function.

9. **Explain the purpose and usage of the `map()` function in Python.**

  Ans. The `map()` function in Python is a built-in function that applies a given function to each item of an iterable (like a list, tuple, or string) and returns an iterator that yields the results.
  
  Purpose:
  
  The primary purpose of `map()` is to avoid explicit loops when you need to apply the same transformation or operation to every element in a sequence. It promotes a more functional programming style, making your code more concise and often more readable.
  
  Usage:
  
  The syntax of `map()` is:
            map(function, iterable, ...)

  - `function`:
    - This is the function that will be applied to each item in the iterable. It can be a built-in function, a user-defined function, or a lambda function.
  - `iterable`:
    - This is the sequence (list, tuple, string, etc.) that we want to apply the function to.
  - `...`(Optional):
    - `map()` can take multiple iterables as arguments. If we provide multiple iterables, the `function` must take as many arguments as there are iterables. The `map()` function will then apply the function to the corresponding elements of each iterable.

10. **What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

  Ans. `map()`, `reduce()`, and `filter()` are all built-in functions in Python that operate on iterables, but they serve distinct purposes:
  
  1. `map()`:
    - Purpose: Applies a given function to each item in an iterable and returns an iterator that yields the results.
    - Transformation: Transforms each element of the iterable.
    - Output: Returns an iterator (which can be converted to a list, tuple, etc.) with the same number of elements as the input iterable.
    - Example:
          numbers = [1, 2, 3, 4, 5]
          squared_numbers = list(map(lambda x: x**2, numbers))
          print(squared_numbers)   # Output: [1, 4, 9, 16, 25]
    
  2. `filter()`:
    - Purpose: Filters elements from an iterable based on a given function (predicate).
    - Selection: Selects elements that satisfy a given condition.
    - Output: Returns an iterator (which can be converted to a list, tuple, etc.) containing only the elements for which the predicate function returns True.
    - Example:
          numbers = [1, 2, 3, 4, 5, 6]
          even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
          print(even_numbers)    # Output: [2, 4, 6]
    
  3. `reduce()`:
    - Purpose: Applies a function cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value.
    - Aggregation: Aggregates elements of the iterable into a single result.
    - Output: Returns a single value.
    - reduce() is part of the functools module in Python, so we need to import it.
    - Example:
          from functools import reduce
          numbers = [1, 2, 3, 4, 5]
          product = reduce(lambda x, y: x * y, numbers)
          print(product)   # Output: 120 (1 * 2 * 3 * 4 * 5)

1. **Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def even_sum(numbers):
  """
  Calculates the sum of all even numbers in a given list.

  Args:
    numbers: A list of numbers.

  Returns:
    The sum of all even numbers in the list.
  """
  sum = 0
  for number in numbers:
    if number % 2 == 0:
      sum += number
  return sum

num1 = [1,3,42,36,2,17,99,104]
result = even_sum(num1)
print(f"The sum of even nos. in the list is: {result}")
```

4. **Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def check_primes_in_range(limit):
    """
    Checks and prints prime numbers from 1 up to a given limit.

    Args:
        limit: The upper limit of the range (inclusive).
    """

    def is_prime(number):
        """Helper function to check if a number is prime."""
        if number <= 1:
            return False
        if number <= 3:
            return True
        if number % 2 == 0 or number % 3 == 0:
            return False
        i = 5
        while i * i <= number:
            if number % i == 0 or number % (i + 2) == 0:
                return False
            i += 6
        return True

    for num in range(1, limit + 1):
        if is_prime(num):
            print(f"{num} is a prime number")

check_primes_in_range(200)
```

5. **Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms**

```python
class FibonacciIterator:
    """
    An iterator that generates the Fibonacci sequence up to a specified number of terms.
    """

    def __init__(self, num_terms):
        """
        Initializes the iterator.

        Args:
            num_terms: The number of Fibonacci terms to generate.
        """
        self.num_terms = num_terms
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        """Returns the iterator object itself."""
        return self

    def __next__(self):
        """
        Generates the next Fibonacci number.

        Raises:
            StopIteration: If the specified number of terms has been reached.

        Returns:
            The next Fibonacci number.
        """
        if self.count < self.num_terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration


n = int(input("Enter the number of terms upto which the Fibonacci Series is to be generated: "))
fib_iter = FibonacciIterator(n)
for num in fib_iter:
    print(num)

```

6. **Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(exponent):
    """
    Generates powers of 2 up to a given exponent.

    Args:
        exponent: The maximum exponent.

    Yields:
        Powers of 2 from 2^0 to 2^exponent.
    """
    for i in range(exponent + 1):
        yield 2 ** i

exponent = int(input("Enter the exponent: "))
for power in powers_of_two(exponent):
    print(power)
```

7. **Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_lines(filepath):
    """
    Reads a file line by line and yields each line as a string.

    Args:
        filepath: The path to the file.

    Yields:
        Each line of the file as a string.
    """
    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Remove trailing newline character
    except FileNotFoundError:
        print(f"Error: File not found at {filepath}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Create a test file
with open("testfile.txt", "w") as f:
    f.write("Line 1\nLine 2\nLine 3")

# Read the file using the generator
for line in read_file_lines("testfile.txt"):
    print(line)

for line in read_file_lines("nonexistentfile.txt"):
    print(line)
```

9. **Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
def celsius_to_fahrenheit(celsius):
  """Converts Celsius to Fahrenheit."""
  return (celsius * 9/5) + 32

temperatures_celsius = [25, 10, 0, -5, 30.5]

# Use map() to convert temperatures
temperatures_fahrenheit = list(map(celsius_to_fahrenheit, temperatures_celsius))

print("Celsius Temperatures:", temperatures_celsius)
print("Fahrenheit Temperatures:", temperatures_fahrenheit)
```

10. **Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def remove_vowels(input_string):
    """
    Removes all vowels from a given string using filter().

    Args:
        input_string: The string to remove vowels from.

    Returns:
        A new string with all vowels removed.
    """

    vowels = "aeiouAEIOU"
    result = "".join(filter(lambda char: char not in vowels, input_string))
    return result


input_string = input("Please enter a word or sentence: ")
result_string = remove_vowels(input_string)
print(f"Original string: {input_string}")
print(f"String without vowels: {result_string}")
```
