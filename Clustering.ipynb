1. What is unsupervised learning in the context of machine learning?

   Ans. Unsupervised learning is a type of machine learning where the model learns from data without explicit guidance. Unlike supervised learning, there are no labeled outputs or target variables. The goal is to find patterns, structures, or relationships within the data itself. Common tasks in unsupervised learning include clustering, dimensionality reduction, and anomaly detection.

2. How does K-Means clustering algorithm work?

   Ans. The K-Means algorithm is an iterative process that aims to partition data points into *k* distinct, non-overlapping clusters. Here's a breakdown of how it works:

   1. **Initialization:** Choose the number of clusters, *k*. Randomly select *k* data points as the initial cluster centroids.
   2. **Assignment:** Assign each data point to the nearest centroid. This forms *k* clusters.
   3. **Update:** Recalculate the centroid for each cluster by taking the mean of all data points assigned to that cluster.
   4. **Iteration:** Repeat steps 2 and 3 until the centroids no longer move significantly, or a maximum number of iterations is reached.

   The algorithm converges when the assignments of data points to clusters no longer change, or when the centroids stabilize.

3. Explain the concept of a dendrogram in hierarchical clustering.

   Ans. A dendrogram is a tree-like diagram that visually represents the hierarchical clustering process. It illustrates the arrangement of clusters produced by the algorithm.

   Here's how to interpret a dendrogram:

   *   **Leaves:** Each leaf node at the bottom of the dendrogram represents a single data point.
   *   **Branches:** Branches connect data points or clusters that are grouped together. The height of a branch represents the distance or dissimilarity between the connected clusters or data points. Shorter branches indicate that the clusters or data points are more similar and were merged earlier in the process.
   *   **Nodes:** Internal nodes represent the merging of two or more clusters.
   *   **Root:** The root of the dendrogram at the top represents a single cluster containing all data points.

   By cutting the dendrogram horizontally at a specific height, we can determine the clusters at a given level of similarity. The number of vertical lines that intersect the horizontal cut indicates the number of clusters at that level. Dendrograms are useful for visualizing the relationships between data points and deciding on the optimal number of clusters.

4. What is the main difference between K-Means and Hierarchical Clustering?

   Ans. The main difference between K-Means and Hierarchical Clustering lies in their approach to forming clusters:

   *   **K-Means Clustering:** This is a **partitioning** method. It requires us to pre-specify the number of clusters (*k*) before the algorithm starts. It aims to partition data points into *k* distinct, non-overlapping clusters by iteratively assigning data points to the nearest centroid and updating the centroids.

   *   **Hierarchical Clustering:** This is a **tree-based** method that builds a hierarchy of clusters. It does not require us to specify the number of clusters beforehand. It can be either **agglomerative** (bottom-up, starting with individual data points and merging them into clusters) or **divisive** (top-down, starting with all data points in one cluster and splitting them). The result is a dendrogram, which allows us to visualize the clustering structure and choose the number of clusters by cutting the dendrogram at a desired level.

   In summary, K-Means creates a flat set of clusters based on a pre-defined number, while Hierarchical Clustering creates a nested sequence of clusters represented by a dendrogram, allowing for flexibility in choosing the number of clusters.

5. What are the advantages of DBSCAN over K-Means?

   Ans. DBSCAN (Density-Based Spatial Clustering of Applications with Noise) offers several advantages over K-Means, particularly in certain scenarios:

   *   **Handles irregularly shaped clusters:** K-Means assumes clusters are spherical and roughly equal in size. DBSCAN, being density-based, can discover clusters of arbitrary shapes.
   *   **Identifies outliers:** DBSCAN explicitly identifies noise points (outliers) that do not belong to any cluster, whereas K-Means forces every data point into a cluster.
   *   **Does not require specifying the number of clusters beforehand:** K-Means requires us to specify the number of clusters (*k*) in advance, which can be challenging if we don't have prior knowledge about the data. DBSCAN determines the number of clusters based on the density of the data.
   *   **More robust to noise:** DBSCAN is less sensitive to noise and outliers compared to K-Means.

   However, DBSCAN has its own limitations, such as difficulty in handling datasets with varying densities and sensitivity to the choice of parameters (epsilon and min_samples).

6. When would you use Silhouette Score in clustering?

   Ans. The Silhouette Score is a metric used to evaluate the quality of clustering results, particularly when we don't have ground truth labels. We would use the Silhouette Score to:

   *   **Determine the optimal number of clusters:** By calculating the Silhouette Score for different values of *k* (number of clusters) in algorithms like K-Means, we can choose the *k* that yields the highest average Silhouette Score.
   *   **Assess the separation distance between the resulting clusters:** A higher Silhouette Score indicates that the data points are well-clustered and are clearly separated from neighboring clusters.
   *   **Evaluate the performance of different clustering algorithms:** We can compare the Silhouette Scores obtained from different clustering algorithms on the same dataset to see which algorithm produces better-defined clusters.

   The Silhouette Score for a single data point measures how similar that data point is to its own cluster compared to other clusters. It ranges from -1 to +1, where:

   *   A score close to +1 indicates that the data point is well inside its own cluster and far from other clusters.
   *   A score close to 0 indicates that the data point is on or near the decision boundary between two neighboring clusters.
   *   A score close to -1 indicates that the data point might have been assigned to the wrong cluster.

   The average Silhouette Score across all data points provides an overall measure of the clustering quality.

7. What are the limitations of Hierarchical Clustering?

   Ans. While Hierarchical Clustering is a powerful technique, it has some limitations:

   *   **Sensitivity to noise and outliers:** Outliers can significantly affect the dendrogram structure and the resulting clusters.
   *   **Computational cost:** For large datasets, hierarchical clustering can be computationally expensive, especially the agglomerative approach, as it requires calculating and storing the distance matrix between all data points. The time complexity is typically O(n^2 log n) or O(n^3), depending on the linkage method.
   *   **Cannot handle large datasets efficiently:** Due to the computational cost, it's not suitable for very large datasets.
   *   **Difficulty in handling different cluster sizes and densities:** Like K-Means, standard hierarchical clustering can struggle with clusters of varying sizes and densities.
   *   **Irreversible decisions:** Once data points are merged or split at a certain level, these decisions cannot be undone in subsequent steps.

Despite these limitations, hierarchical clustering is valuable for understanding the structure of the data and visualizing the relationships between data points through the dendrogram.

8. Why is feature scaling important in clustering algorithms like K-Means?

   Ans. Feature scaling is crucial in clustering algorithms like K-Means because these algorithms often rely on distance metrics (e.g., Euclidean distance) to determine the similarity or dissimilarity between data points. If the features have different scales, features with larger values will dominate the distance calculation, effectively masking the influence of features with smaller values.

   Here's why it's important:

   *   **Distance-based algorithms:** Algorithms like K-Means group data points based on their proximity in the feature space. If features are not scaled, a feature with a large range of values can disproportionately influence the distance calculation compared to a feature with a small range.
   *   **Equal contribution of features:** Scaling ensures that all features contribute equally to the distance calculation, regardless of their original scale. This prevents features with larger magnitudes from having an undue influence on the clustering results.
   *   **Improved performance:** Scaling can lead to better clustering performance by ensuring that the algorithm accurately captures the true relationships between data points based on all relevant features.

   Common feature scaling techniques include:

   *   **Standardization (Z-score normalization):** Transforms the data to have zero mean and unit variance.
   *   **Normalization (Min-Max scaling):** Scales the data to a fixed range, typically between 0 and 1.

   By applying feature scaling, we ensure that the clustering algorithm considers all features appropriately, leading to more meaningful and accurate clusters.

9. How does DBSCAN identify noise points?

   Ans. In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), noise points (outliers) are identified based on their density relative to their neighbors. Here's how it works:

   DBSCAN requires two parameters:

   *   **epsilon (ε):** The maximum distance between two samples for one to be considered as in the neighborhood of the other.
   *   **min_samples:** The number of samples in a neighborhood for a point to be considered as a core point.

   DBSCAN categorizes points into three types:

   *   **Core point:** A point is a core point if there are at least `min_samples` (including the point itself) within a distance of `epsilon` from it.
   *   **Border point:** A point is a border point if it is not a core point but is in the neighborhood of a core point.
   *   **Noise point:** A point is a noise point if it is neither a core point nor a border point. This means that within its `epsilon` neighborhood, there are fewer than `min_samples` points, and it is not reachable from any core point.

   Essentially, noise points are those that reside in low-density regions of the data space and are not close enough to any dense cluster (represented by core points) to be included. DBSCAN naturally identifies these isolated points as noise, without forcing them into a cluster.

10. Define inertia in the context of K-Means.

    Ans. In the context of K-Means clustering, **inertia** (also known as within-cluster sum of squares) is a metric used to evaluate the quality of the clustering. It measures the sum of squared distances of samples to their closest cluster center.

    More formally, for a set of data points X and a set of cluster centroids C, inertia is calculated as:

    Inertia = Σ from {i=1} to n * Σ j ε k * ||x_i - c_j||²

    where:
    *   n is the number of data points.
    *   k is the number of clusters.
    *   x_i is the i-th data point.
    *   c_j is the centroid of the j-th cluster.
    *   || x_i - c_j ||² is the squared Euclidean distance between the data point x_i and the centroid c_j.

    A lower inertia value generally indicates better clustering, as it means the data points are closer to their respective cluster centroids. However, inertia tends to decrease as the number of clusters (*k*) increases, so it's important to use it in conjunction with other metrics or methods (like the elbow method) to determine the optimal number of clusters.

11. What is the elbow method in K-Means clustering?

    Ans. The elbow method is a heuristic used to determine the optimal number of clusters (*k*) for K-Means clustering. It is based on the idea that as we increase the number of clusters, the inertia (within-cluster sum of squares) will decrease. However, the rate of decrease typically slows down after a certain point, creating an "elbow" shape in the plot of inertia versus the number of clusters.

    Here's how it works:

    1.  Run K-Means clustering for a range of *k* values (e.g., from 1 to 10).
    2.  For each *k*, calculate the inertia.
    3.  Plot the inertia values against the number of clusters (*k*).
    4.  Look for the "elbow" point in the plot, which is the point where the rate of decrease in inertia sharply changes. This point is often considered the optimal number of clusters.

    The idea is that adding more clusters beyond the elbow point provides diminishing returns in terms of reducing within-cluster variation. While the elbow method is a useful visual aid, it can sometimes be subjective, and other metrics like the Silhouette Score can also be used to evaluate the optimal number of clusters.

12. What is "density" in DBSCAN?

    In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), "density" refers to the concentration of data points in a given region of the data space. The algorithm identifies clusters as areas of high density separated by areas of low density.

    The concept of density is defined by two parameters:

    1.  **Epsilon (ε):** This is a distance threshold. A data point's "neighborhood" is defined as all other data points within a distance of ε from it.
    2.  **Min_samples:** This is a count threshold. For a region to be considered "dense," it must contain at least `min_samples` within the ε-neighborhood of a given point.

    Based on these parameters, DBSCAN classifies data points into three types:

    *   **Core points:** These are points that have at least `min_samples` within their ε-neighborhood (including themselves). They are located in the interior of a dense region.
    *   **Border points:** These are points that have fewer than `min_samples` within their ε-neighborhood but are within the ε-neighborhood of a core point. They are on the edge of a dense region.
    *   **Noise points:** These are points that are neither core points nor border points. Their ε-neighborhood contains fewer than `min_samples`, and they are not reachable from any core point. These points are considered outliers or noise.

    DBSCAN groups together core points that are reachable from each other (directly or indirectly through other core points) and includes their border points in the cluster. Regions with insufficient density (those containing only noise points) are not considered clusters.

    Therefore, "density" in DBSCAN is a measure of how many data points are packed together within a specified radius (ε), and this density is used to define and discover clusters of arbitrary shapes, while also identifying outliers.

13. Can hierarchical clustering be used on categorical data?

    Yes, hierarchical clustering can be used on categorical data, but it's not as straightforward as using it with numerical data. Standard distance metrics like Euclidean distance, which are commonly used in hierarchical clustering, are not directly applicable to categorical variables. Therefore, we need to preprocess the categorical data to represent it in a way that allows for meaningful distance calculations.

    Here are common approaches to use hierarchical clustering with categorical data:

    *   **Convert categorical data to numerical data:**
        *   **One-Hot Encoding:** This is a common technique where each category of a feature is converted into a new binary feature (0 or 1). While this allows for using distance metrics, it can significantly increase the dimensionality of the data, especially if we have features with many categories.
        *   **Ordinal Encoding:** If the categorical data has a natural order (e.g., "small," "medium," "large"), we can assign numerical values based on this order. However, this assumes a specific relationship between categories that might not always exist.
    *   **Use appropriate distance metrics:** Once the data is converted to a numerical format (like after one-hot encoding), we can use distance metrics suitable for binary data, such as:
        *   **Jaccard Distance:** This is often used for binary data and measures the dissimilarity between two sets.
        *   **Hamming Distance:** This measures the number of positions at which the corresponding symbols are different.
    *   **Specialized distance metrics for categorical data:** Some distance metrics are designed specifically for categorical data, such as the Gower distance, which can handle a mix of categorical and numerical features.

    After preprocessing the data and choosing an appropriate distance metric, we can apply hierarchical clustering as we would with numerical data. The choice of preprocessing method and distance metric depends on the nature of our categorical data and the goals of our clustering analysis.

15. Explain the term "linkage criteria" in hierarchical clustering.

    In hierarchical clustering, "linkage criteria" (or linkage methods) determine how the distance between clusters is calculated. Since hierarchical clustering involves merging or splitting clusters based on their proximity, the way this distance is defined significantly impacts the resulting dendrogram and the structure of the clusters.

    Here are some common linkage criteria:

    *   **Single Linkage:** The distance between two clusters is the minimum distance between any single data point in one cluster and any single data point in the other cluster. This method tends to create long, "chain-like" clusters.
    *   **Complete Linkage:** The distance between two clusters is the maximum distance between any single data point in one cluster and any single data point in the other cluster. This method tends to create more compact, spherical clusters.
    *   **Average Linkage:** The distance between two clusters is the average distance between all pairs of data points, where one point is from each cluster. This method is a compromise between single and complete linkage and is less sensitive to outliers than single linkage.
    *   **Centroid Linkage:** The distance between two clusters is the distance between their centroids (the mean of all data points in each cluster).
    *   **Ward's Method:** This method minimizes the variance of the clusters being merged. It tends to produce clusters of roughly equal size and is often considered effective, although it is sensitive to the initial scaling of the data.

    The choice of linkage criterion depends on the nature of the data and the desired shape and characteristics of the clusters.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

    K-Means clustering can perform poorly on data with varying cluster sizes or densities primarily because of its underlying assumptions and mechanism:

    *   **Assumption of Spherical and Equally Sized Clusters:** K-Means implicitly assumes that clusters are roughly spherical and of similar size. It tries to partition the data space into Voronoi cells around the centroids. When clusters have significantly different sizes or are non-spherical (e.g., elongated or irregular shapes), the algorithm's distance-based assignments can lead to misclassifications, particularly at the boundaries between clusters. It tends to assign points to the nearest centroid, even if that centroid belongs to a much denser or larger cluster that the point isn't truly part of.

    *   **Reliance on Mean (Centroid):** The algorithm uses the mean of the data points to calculate the cluster centroid. The mean is sensitive to outliers and can be heavily influenced by larger or denser clusters, pulling the centroid away from the true center of smaller or less dense clusters. This can cause data points from smaller/less dense clusters to be incorrectly assigned to larger/denser ones.

    *   **Equal Weighting of Features:** K-Means uses Euclidean distance (or similar metrics) which treats all features equally. If clusters differ in density due to variations in the spread of data along different features, K-Means doesn't inherently account for this, potentially leading to poor separation.

    In contrast, density-based algorithms like DBSCAN are better suited for discovering clusters of arbitrary shapes and handling varying densities because they define clusters based on the concentration of data points rather than distances to a central point.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

    DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has two core parameters that significantly influence the clustering results:

    *   **Epsilon (ε):** This parameter defines the radius of the neighborhood around a data point. It determines how far the algorithm looks to find neighboring points.
        *   **Influence:** A smaller ε value means that only points very close to each other are considered neighbors. This can lead to more clusters and potentially identify finer-grained structures, but it might also split genuinely dense regions into multiple clusters if the points are not extremely close. A larger ε value considers points further away as neighbors. This can result in fewer, larger clusters and might merge regions that should be separate. The choice of ε is crucial and often determined through exploration or domain knowledge.

    *   **Min_samples:** This parameter defines the minimum number of points required in a neighborhood for a point to be considered a core point (including the point itself).
        *   **Influence:** A larger `min_samples` value makes the density requirement stricter. This means that only very dense regions will be considered clusters, and more points will be classified as noise. This can help in identifying more significant clusters and filtering out smaller, less dense groups. A smaller `min_samples` value makes the density requirement less strict, allowing smaller or less dense regions to form clusters. This can lead to more clusters being identified, but it might also be more sensitive to noise.

    Both ε and `min_samples` are interdependent. For instance, a smaller ε might require a smaller `min_samples` to find core points in a given density, while a larger ε might require a larger `min_samples` to maintain a similar density threshold. The optimal values for these parameters often require experimentation and understanding of the dataset's characteristics.

18. How does K-Means++ improve upon standard K-Means initialization?

    K-Means++ is an initialization method for the K-Means algorithm that aims to select initial cluster centroids that are spread out across the data space. This improves upon the standard K-Means initialization, which typically involves randomly selecting *k* data points as initial centroids.

    The problem with random initialization in standard K-Means is that it can sometimes lead to:

    *   **Suboptimal Solutions:** If the initial centroids are poorly chosen (e.g., all clustered together), the K-Means algorithm may converge to a local optimum rather than the global optimum, resulting in suboptimal clustering.
    *   **Sensitivity to Initial Centroids:** The final clustering result can be highly dependent on the initial random selection of centroids.

    K-Means++ addresses these issues by using a probabilistic approach to select the initial centroids. Here's a simplified breakdown of the K-Means++ initialization process:

    1.  Select the first centroid randomly from the data points.
    2.  For each remaining data point, calculate the distance to the nearest already selected centroid.
    3.  Select the next centroid from the remaining data points with a probability proportional to the square of its distance to the nearest already selected centroid. This means points that are farther away from existing centroids have a higher chance of being selected as the next centroid.
    4.  Repeat steps 2 and 3 until *k* centroids have been selected.

    By preferentially selecting points that are far from existing centroids, K-Means++ ensures that the initial centroids are more spread out, covering the data space more effectively. This significantly reduces the likelihood of converging to a poor local optimum and generally leads to better clustering results compared to standard random initialization.

    In essence, K-Means++ provides a more robust and effective way to start the K-Means algorithm, leading to more consistent and often better quality clusters.

19. What is agglomerative clustering?

    Agglomerative clustering is a type of hierarchical clustering that follows a "bottom-up" approach. It starts with each data point as its own individual cluster and iteratively merges the closest pairs of clusters until only one cluster remains (or a desired number of clusters is reached).

    Here's a breakdown of the process:

    1.  **Initialization:** Each data point is considered a single cluster.
    2.  **Calculate Distances:** Compute the pairwise distances between all clusters. The distance between clusters is determined by a chosen linkage criterion (e.g., single linkage, complete linkage, average linkage, Ward's method).
    3.  **Merge Clusters:** Merge the two closest clusters based on the calculated distances and the linkage criterion.
    4.  **Update Distances:** Recalculate the distances between the new cluster and all other existing clusters.
    5.  **Repeat:** Repeat steps 3 and 4 until only one cluster remains or a stopping condition is met (e.g., reaching a specific number of clusters or a certain distance threshold).

    The process of merging clusters can be visualized using a dendrogram, where each merge is represented by a node, and the height of the node indicates the distance at which the clusters were merged.

    Agglomerative clustering is useful for exploring the hierarchical structure of data and does not require specifying the number of clusters beforehand. However, it can be computationally expensive for large datasets due to the need to calculate and update the distance matrix.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

    While inertia is a useful metric for evaluating K-Means clustering, the Silhouette Score is often considered a better metric for model evaluation, particularly when comparing different numbers of clusters or different clustering algorithms. Here's why:

    *   **Accounts for both intra-cluster cohesion and inter-cluster separation:** Inertia only measures how close data points are to their own cluster centroid (intra-cluster cohesion). It doesn't consider how well separated the clusters are from each other. The Silhouette Score, on the other hand, takes into account both the average distance to points in its own cluster (cohesion) and the average distance to points in the nearest neighboring cluster (separation). This provides a more comprehensive evaluation of the clustering quality.

    *   **Provides a standardized score:** The Silhouette Score is normalized and ranges from -1 to +1. This makes it easier to interpret and compare across different datasets or clustering results. A score close to +1 indicates well-separated clusters, a score near 0 indicates overlapping clusters, and a score close to -1 suggests that data points might have been assigned to the wrong cluster. Inertia, on the other hand, is a sum of squared distances and its value depends on the scale of the data and the number of data points, making direct comparisons difficult.

    *   **Less sensitive to the number of clusters:** Inertia inherently decreases as the number of clusters increases, even if the additional clusters don't represent meaningful structures in the data. This makes it difficult to use inertia alone to determine the optimal number of clusters (though the elbow method tries to address this). The Silhouette Score, while not perfect, is less prone to this issue and can provide a more reliable indication of the optimal number of clusters by looking for the highest average Silhouette Score.

    In summary, the Silhouette Score offers a more balanced and interpretable measure of clustering quality by considering both how tightly data points are grouped within clusters and how well the clusters are separated from each other. This makes it a more robust metric for evaluating and comparing different clustering solutions.

21. **Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.**

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data with 4 centers
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

44. **Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.**

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply K-Means clustering (let's use 3 clusters, matching the known classes)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

# Calculate the Silhouette Coefficient for each sample
sample_silhouette_values = silhouette_samples(X, labels)

# Plot the Silhouette Coefficient for each sample
plt.figure(figsize=(10, 6))
y_lower = 10 # Starting point for the first silhouette plot

for i in range(3): # For each cluster
    # Aggregate the silhouette scores for samples belonging to cluster i, and sort them
    ith_cluster_silhouette_values = sample_silhouette_values[labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.viridis(float(i) / 3)
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)

    # Label the silhouette plots with their cluster numbers at the middle
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    # Compute the new y_lower for next plot
    y_lower = y_upper + 10 # 10 for the 0 samples

plt.title("Silhouette plot for the various clusters")
plt.xlabel("The silhouette coefficient values")
plt.ylabel("Cluster label")

# The vertical line for average silhouette score of all the values
average_score = np.mean(sample_silhouette_values)
plt.axvline(x=average_score, color="red", linestyle="--", label=f'Average Silhouette Score: {average_score:.2f}')
plt.yticks([])  # Clear the y-axis labels / ticks
plt.xticks(np.arange(-1, 1.1, 0.1))
plt.legend()
plt.show()

print(f"Average Silhouette Score: {average_score}")
```
